## Основные теоретические положения

### Виды детерминированных дискретных сигналов

#### Единичный цифровой импульс

$$\delta_d(k)=
\begin{cases}
1, & k=0,\\
0, & k\neq0.
\end{cases}
,n=...,-2,-1,0,1,2,...$$
Используется как аналог дельта-функции:
$$\delta(t)=
\begin{cases}
\infty, & t=0,\\
0, & t\neq0.
\end{cases}
,\int_{-\infty}^{\infty}\delta(t)dt=1$$

#### Дискретный единичный скачок:
$$\sigma_d(k)=
\begin{cases}
1, & k\geq0,\\
0, & k<0.
\end{cases}
,k=...,-2,-1,0,1,2,...$$
Используется как дискретный аналог функции единичного скачка:
$$\sigma(t)=
\begin{cases}
1, & t>0,\\
0.5 & \text{или неопределено,} & t=0, \\
0, & t<0.
\end{cases}$$
#### Дискретная экспоненциальная функция
$$s(k)=
\begin{cases}
0, & k<0,\\
a^k, & k\geq0.
\end{cases}
,k=...,-2,-1,0,1,2,...$$
#### Дискретная затухающая синусоида:
$$s(k)=a^kcos(k\omega+\varphi)$$
#### Дискретный комплексный гармонический сигнал:
$$s_2(n)=Cexp(j\hat\omega_0k)$$

## Выполнение работы

### 1. Моделирование еденичного цифрового импульса.

```python
import math
from matplotlib import pyplot as plt

# Исходные данные
Nb = 5
N = 30 + Nb % 5
T = 0.0005 * (1 + Nb % 3)
a = (-1) ** Nb * (0.8 + 0.005 * Nb)
C = 1 + Nb % 5
freq = math.pi / (6 + Nb % 5)
m = 5 + Nb % 5

# Единичный импульс
def unit_impulse(k):
    return 1 if k == 0 else 0

# Время: дискретное и дискретное нормированное
discrete_time = [n * T for n in range(N)]
normalized_discrete_time = [n for n in range(N)]

dt_impulse = [unit_impulse(t) for t in discrete_time]
ndt_impulse = [unit_impulse(t) for t in normalized_discrete_time]

def show_plot(X, Y):
    plt.clf()
    plt.grid()
    plt.stem(X, Y, linefmt = 'C1-')
    plt.show()
    
show_plot(discrete_time, dt_impulse)
show_plot(normalized_discrete_time, ndt_impulse)
```

#### a. Взаимосвязь между дискретным и дискретным нормированным временем.
$nT$ - дискретное время, где $T$ - период дискретизации. 
$n$ - дискретное нормированное время, где $Т = 1$  
#### b. Различие между единичным цифровым импульсом и дельта-функцией. 
Дельта-функция в точке 0 равна бесконечности, в то время как единичный цифровой импульс в точке 0 равен 1.

### 2. Моделирование дискретного еденичного скачка.

```python
def unit_step(k): # Единичный скачок
    return 1 if k >= 0 else 0

dt_step = [unit_step(t) for t in discrete_time]
ndt_step = [unit_step(t) for t in normalized_discrete_time]

show_plot(discrete_time, dt_step)
show_plot(normalized_discrete_time, ndt_step)
```

#### а. Соответствие между дискретным и аналоговым единичными скачками.
Дискретная функция единичного скачка равна 1 в точке 0. Дискретная функция определена на множестве целых чисел.  
Аналоговая фукнция единичного скачка равна 0.5 или неопределена.  Аналоговая функция определена на множестве действительных чисел.  
#### b.Чему равна частота дискретизации дискретного единичного скачка.
Частота дискретизации единичного скачка равна $f_d = \frac{1}{T}$. Высчитаем её:

```python
1/T
```

### 3. Моделирование дискретной экспоненциальной функции.

```python
import numpy
def discrete_exp(k): # Дискретная экспоненциальная функция
    return 0 if k < 0 else a ** k

dt_exp = [discrete_exp(t) for t in discrete_time]
ndt_exp = [discrete_exp(t) for t in normalized_discrete_time]

show_plot(discrete_time, numpy.real(dt_exp))
show_plot(normalized_discrete_time, numpy.real(ndt_exp))
```

### 4. Моделирование дискретного комплексного гармонического сигнала.

```python
def discrete_harmonic(k): # Дискретная гармоническая функция
    return C * numpy.exp(1j * freq * k)

ndt_dt_harmonicexp = [discrete_harmonic(t) for t in normalized_discrete_time]

show_plot(normalized_discrete_time, numpy.real(ndt_dt_harmonicexp))
show_plot(normalized_discrete_time, numpy.imag(ndt_dt_harmonicexp))
```

$C \cdot e^{j \hat{\omega}_0 k} = C \cdot cos(\hat{\omega}_0 k) + j \cdot C \cdot sin(\hat{\omega}_0 k)$ - формула Эйлера  
$C \cdot cos(\hat{\omega}_0 k)$ - действительная часть  
$C \cdot sin(\hat{\omega}_0 k)$ - мнимая часть

### 5. Вывод графиков задержанных последовательностей.

```python
# Сдвиг последовательности на m
delayed_discrete_time = [(k - m) for k in range(N)]

# Задержанные графики
delayed_impulse = [unit_impulse(t) for t in delayed_discrete_time]
delaed_step = [unit_step(t) for t in delayed_discrete_time]
delayed_exp = [discrete_exp(t) for t in delayed_discrete_time]

show_plot(discrete_time, delayed_impulse)
show_plot(discrete_time, delaed_step)
show_plot(discrete_time, delayed_exp)
```

Внесём изменения в функции сигналов, заменив аргумент **k** на **k - m** (смещая график по оси X на m единиц дискретного времени вправо):

Единичный цифровой импульс:
$$\delta_d(k)=
\begin{cases}
1, & k-m=0,\\
0, & k-m\neq0.
\end{cases}
$$
Дискретный единичный скачок:
$$\sigma_d(k)=
\begin{cases}
1, & k-m\geq0,\\
0, & k-m<0.
\end{cases}
$$
Дискретная экспоненциальная функция:
$$s(k)=
\begin{cases}
0, & k-m<0,\\
a^{(k-m)}, & k-m\geq0.
\end{cases}
$$
