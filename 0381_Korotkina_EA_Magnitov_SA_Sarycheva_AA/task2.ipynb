## Основные теоретические положения

**Дискретный единичный скачок:**

$$\sigma_d (k) = \begin{equation*}
\begin{cases}
   1, k \geq 0\\
   0, k < 0
 \end{cases}
\end{equation*} \quad k=...,-2,-1,0,1,2, ...$$


**Линейная комбинация** - выражение, построенное на множестве элементов путём умножения каждого элемента на коэффициенты с последующим сложением результатов.

**Дискретный прямоугольный импульс:**
$$
s_1(k) = \begin{equation*}
\begin{cases}
   U, n_0 \leq k \leq n_0 + n_{imp} - 1\\
   0, иначе
 \end{cases}
\end{equation*}
$$
**Линейная комбинация дискретных гармонических сигналов:**
$$s_2(k) = a_1x_1(k) + a_2 x_2 (k) + a_3 x_3 (k), \space x_i(k) = B_i\sin(\hat\omega_i k), i = 1,2,3$$

**Характеристики сигналов:**

1) Среднее значение: $M=\frac{\sum\limits_{i=0}^{N-1} s_2(i)}{N}$

2) Энергия: $E=\sum\limits_{i=0}^{N-1} s_2^2(i)$

3) Средняя мощность: $P=\frac{E}{N}$

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

Инициализация начальных данных.

```python
Nb = 6
N = 30 + Nb % 5
a = (-1) ** Nb * (0.8 + 0.005 * Nb)
C = 1 + Nb % 5
w0 = math.pi / (6 + Nb % 5)
U = Nb
n0 = Nb % 5 + 3
n_imp = Nb % 5 + 5
B1 = 1.5 + Nb % 5
B2 = 5.7 - Nb % 5
B3 = 2.2 + Nb % 5
w1 = math.pi / (4 + Nb % 5)
w2 = math.pi / (8 + Nb % 5)
w3 = math.pi / (16 + Nb % 5)
a1 = 1.5 - Nb % 5
a2 = 0.7 + Nb % 5
a3 = 1.4 + Nb % 5
```

Инициализация функций для линейной комбинации.

```python
def x1(x):
  return B1*np.sin(w1*x)

def x2(x):
  return B2*np.sin(w2*x)

def x3(x):
  return B3*np.sin(w3*x)
```

Инициализация дискретного времени и нормированного дискретного времени.

```python
norm_discrete_time_points = np.linspace(0, N - 1, N)
task2_time_points = np.linspace(0, 5*N - 1, 5*N)
task4_time_points = np.linspace(0, 2*N - 1, 2*N)
```

**1. Моделирование дискретного прямоугольного импульса $s_1(k)$ на основе дискретного единичного скачка $\sigma_d (k)$ с выводом графика на интервале времени $n \in [0,N-1]$.**

```python
def square_pulse(time_points, moment):
    y = np.zeros(len(time_points))
    for i in range(len(y)):
      if (moment <= i and i <= moment + n_imp - 1):
        y[i] = U
    return y
```

```python
plt.figure(figsize=(18, 5))
plt.stem(norm_discrete_time_points, square_pulse(norm_discrete_time_points, n0), linefmt='green')
plt.title("Дискретный прямоугольный импульс S1")
plt.show()
```

Пояснение:

Дискретный прямоугольный импульс можно составить из двух единичных скачков одинаковой амплитуды. Один скачок должен произойти в момент времени $n_0$, а второй - в момент времени $n_0+n_{imp}-1$. Так как сигналы $\sigma_d (k)$ больше не обнулятся, то при вычитании второго из первого вся временная область после $n_0+n_{imp}-1$ обратится в 0, оставив прямоугольный испульс в нужной области:
$$s_1(k)=U\cdot (\sigma_d (k-n_0) - \sigma_d (k-n_0-n_{imp}+1))$$

**2. Моделирование линейной комбинации дискретных гармонических сигналов $s_2(k)$, используя $x_i(k) = B_i\sin(\hat\omega_i k)$, $i = 1,2,3$, с выводом графиков на интервале времени $n \in [0, 5N - 1]$.**

```python
def linear_combination(time_points):
  y = np.zeros(len(time_points))
  for i in range(len(time_points)):
    y[i] = a1*x1(time_points[i]) + a2*x2(time_points[i]) + a3*x3(time_points[i])
  return y
```

```python
plt.figure(figsize=(18, 5))
plt.stem(task2_time_points, linear_combination(task2_time_points), linefmt='green')
plt.title("Линейная комбинация S2")
plt.show()
```

```python
plt.figure(figsize=(18, 5))
plt.stem(task2_time_points, x1(task2_time_points), linefmt='blue')
plt.title("Гармонический сигнал X1")
plt.show()
```

```python
plt.figure(figsize=(18, 5))
plt.stem(task2_time_points, x2(task2_time_points), linefmt='blue')
plt.title("Гармонический сигнал X2")
plt.show()
```

```python
plt.figure(figsize=(18, 5))
plt.stem(task2_time_points, x3(task2_time_points), linefmt='blue')
plt.title("Гармонический сигнал X3")
plt.show()
```

```python
#Среднее значение
mean_s2 = sum([linear_combination(task2_time_points)[i] for i in range(len(task2_time_points))])/(5*N)

#Энергия
E = sum(linear_combination(task2_time_points)**2)

#Средняя мощность
P = E/(5*N)

print('Среднее значение (mean_s2) = %f, Энергия (E) = %f, Средняя мощность (P) = %f.' % (mean_s2, E, P))
```

Пояснение:

A. Так как сигнал представляет собой линейную комбинацию, то он получается путем сложения сигналов $x_1(t), x_2(t), x_3(t)$ и умножения каждого слагаемого на соответсвенный численный коэффициент. При таком задании $s_2(t)$ сигнал полностью удовлетворяет определению линейной комбинации, что позволяет найти найти значение $s_2(t)$ в момент времени $t_0$ через подстановку значения в известные $x_1(t), x_2(t), x_3(t)$.

B. Характеристики:

Среднее значение:
$$M=\frac{\sum\limits_{i=0}^{5\cdot N-1} s_2(i)}{5\cdot N}=0.825$$
Энергия:
$$E=\sum\limits_{i=0}^{5\cdot N-1} s_2^2(i)=9493.592$$
Средняя мощность:
$$P=\frac{E}{5\cdot N}=61.249$$


**3. Визуализация дискретного сигнала $s_3(k)$, представляющего собой дискретный гармонический сигнал $s(k)$ с экспоненциальной огибающей $|a|^k$, на интервале времени $n \in [0, N-1]$.**

```python
def discrete_signal_s3(time_points):
  y = np.zeros(len(time_points))
  for i in range(len(time_points)):
    y[i] = C * np.abs(a) ** time_points[i] * np.cos(w0*time_points[i])
  return y

def dop_function(time_points):
  y = np.zeros(len(time_points))
  for i in range(len(time_points)):
    y[i] = C* np.abs(a) ** time_points[i]
  return y

def dop_dis_function(time_points):
  y = np.zeros(len(time_points))
  for i in range(len(time_points)):
    y[i] = C * np.cos(w0*time_points[i])
  return y
```

```python
plt.figure(figsize=(18, 5))
plt.stem(norm_discrete_time_points, discrete_signal_s3(norm_discrete_time_points), linefmt='green')
plt.title("Дискретный гармонический сигнал с экспоненциальной огибающей S3")
plt.show()
```

```python
plt.figure(figsize=(18, 5))
plt.stem(norm_discrete_time_points, dop_dis_function(norm_discrete_time_points), linefmt='lawngreen')
plt.plot(norm_discrete_time_points, dop_function(norm_discrete_time_points), color="cornflowerblue")
plt.stem(norm_discrete_time_points, discrete_signal_s3(norm_discrete_time_points), linefmt='green')
plt.title("Влияние огибающей на гармонический сигнал")
plt.show()
```

Пояснение:

Аналитическое представление сигнала описывается формулой:
$$s_3(k) = C \cdot |a|^k \cdot cos(\hat{ω}_0k)$$
т.е. дискретный гармонический сигнал ограничивается экспоненциальной огибающей. На графике продемонстрированы и дискретный сигнал, его огибающая и сигнал без влияния огибающей. Это визуально показывает, как сигнал меняет свои значения и меняется под влиянием огибающей.

**4. Визуализация пяти периодов периодической последовательности $s_4(k)$ дискретных прямоугольных импульсов амплитуды $U$ и длительности $n_{imp}$ с периодом, вдвое большим длительности импульса.**

```python
def five_period_signal(time_points):
  y = np.zeros(len(time_points))
  for i in range(0, 5):
    y += square_pulse(time_points, n0 + 2*n_imp*i)
  return y
```

```python
plt.figure(figsize=(18, 5))
plt.stem(task4_time_points, five_period_signal(task4_time_points), linefmt='green')
plt.title("Сигнал периодической последовательности S4")
plt.show()
```

Пояснение:

Моделирование $N$ прямоугольных импульсов с периодом $T$ можно охарактеризовать формулой:

$$s_4(k) = \sum\limits_{i=0}^{N-1} s_1(k-iT)$$

которая обозначает следующее: $N$ отдельно сгенерированных импульсов с разной задержкой, зависящей от периода $T$, накладываются на одну временную ось, откуда и получается периодическая последовательность $s_4(k)$.
