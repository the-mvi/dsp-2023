```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

1. Выведем формулы для передаточной функции нерекурсивного фильтра, соответствующего сглаживанию прямой линией по 3, 5, 7 и 9 точкам.

Входной сигнал: $s(t)$

Выходной сигнал: $y(t) = A + Bt$.

Приближение (в смысле МНК) прямой линией по пяти точкам:

$F(A, B) = \sum\limits_{k=-m}^{m} (s_k - y_k)^2 = \sum\limits_{k=-m}^{m} (s_k - A - Bk)^2 \Rightarrow min$

Посчитаем частные производные по A и B, получим:

$\begin{equation*}
 \begin{cases}
   {{\delta F(A, B)} \over {\delta A}} = 0\\
   {{\delta F(A, B)} \over {\delta B}} = 0
 \end{cases}
\end{equation*} \Leftrightarrow$

$\Leftrightarrow \begin{equation*}
 \begin{cases}
   -2 \sum\limits_{k=-m}^{m} (s_k - A - Bk) = 0\\
   -2 \sum\limits_{k=-m}^{m} (ks_k - kA - Bk^2) = 0
 \end{cases}
\end{equation*} \Leftrightarrow$



Тогда система нормальных уравнений:

$\begin{equation*}
 \begin{cases}
   A = {1 \over {2m + 1}} \sum\limits_{k=-m}^{m} s_k \\
   B = {{\sum\limits_{k=-m}^{m} s_k} \over \sum\limits_{k=-m}^{m} k^2}
   \end{cases}
\end{equation*}$

В итоге получаем:

$y_0 = A = {1 \over {2m + 1}}\sum\limits_{k=-m}^{m} s_k = {1 \over {2m + 1}}\sum\limits_{k=-m}^{m} (s_{-m} + s_{-m+1} + ... + s_{m-1} + s_m)$

В общем случае:

$y_n = {1 \over {2m + 1}}\sum\limits_{k=-m}^{m} s_k = {1 \over {2m + 1}}\sum\limits_{k=-m}^{m} (s_{-m+n} + s_{-m+1+n} + ... + s_{m-1+n} + s_{m+n})$

$s_n = e^{i \omega n}$

$y_n = {1 \over {2m + 1}}\sum\limits_{k=-m}^{m} (e^{-m i \omega} + e^{(-m+1)i \omega} + ... + e^{(m-1)i \omega} + e^{m i \omega}) = H(\omega)$

$H(\omega) = {1 \over {2m + 1}}\sum\limits_{k=-m}^{m} (1 + 2\cos(\omega) + 2\cos(2\omega) + ... + 2\cos(m\omega))$

$H(\omega) = {{\sin({{(2m + 1)\omega} \over 2})} \over {(2m+1)\sin({\omega \over 2})}}$

$H(\omega) = H(2\pi f) = \tilde{H}(f)$

$H_k(f) = {{\sin(k\pi f)} \over {k sin(\pi f)}}$

Формула для передаточной функции нерекурсивного фильтра, соответствующего сглаживанию прямой линией по 3 точкам:

$H_3(f) = {{\sin(3\pi f)} \over {3 sin(\pi f)}}$

Аналогично для 5, 7, 9 точек:

$H_5(f) = {{\sin(5\pi f)} \over {5 sin(\pi f)}}$

$H_7(f) = {{\sin(7\pi f)} \over {7 sin(\pi f)}}$

$H_9(f) = {{\sin(9\pi f)} \over {9 sin(\pi f)}}$

Построим графики $\tilde{H}(f)$:

```python
def h_k(k, f):
    return np.sin(k*math.pi*f)/(k*np.sin(f*math.pi))

f = np.linspace(0.001, 0.5, 64)
for k in range(3, 10, 2):
    plt.plot(f, h_k(k, f), label=r'$H_{}$'.format(k))
plt.xlabel('t')
plt.ylabel(r'$H(t)$')
plt.legend()
plt.title("Графики для передаточных функций")
plt.show()
```

Можно заметить, что за один период каждая из передаточных функций имеет ровно столько экстремумов, столько было использовано точек для ее построения. Получается, чем больше экстремумов, тем больше колебаний. К тому же порядок гладкости увеличивается с увеличением количества точек для построения функции.

2. Выведем формулы для передаточной функции нерекурсивного фильтра, соответствующего сглаживанию полиномом второй степени по 7, 9, 11 и 13 точкам.

Посчитаем частные производные по A и C, получим:

$\begin{equation*}
 \begin{cases}
   {{\delta F(A, B, C)} \over {\delta A}} = 0\\
   {{\delta F(A, B, C)} \over {\delta C}} = 0

$\begin{equation*}

\end{equation*}$

Подставим ${{m(m+1)(2m+1)} \over 3} C$ во второе уравнение:

Тогда:

$A = {{\sum\limits_{k=-m}^{m} k^2 s_k - {{3m^2 + 3m - 1} \over 5} \sum\limits_{k=-m}^{m} s_k} \over {{{m(m+1)(2m+1)} \over 3} - {{(3m^2 + 3m - 1)(2m + 1)} \over 5}}}$

В итоге получаем:

$y_n = {1 \over 21} (-2e^{-3i \omega} + 3e^{-2i \omega} + 6e^{-i \omega} + 7 + 6e^{i \omega} + 3e^{2i \omega} - 2e^{3i \omega}) e^{i\omega n} = H(\omega)e^{i\omega n}$

$H(\omega) = {1 \over 21} (-2e^{-3i \omega} + 3e^{-2i \omega} + 6e^{-i \omega} + 7 + 6e^{i \omega} + 3e^{2i \omega} - 2e^{3i \omega})$

$H(\omega) = {1 \over 21} (7 + 12\cos{\omega} + 6\cos{2\omega} - 4\cos{3\omega})$

$H(\omega) = H(2\pi f) = \tilde{H}(f)$

Формула для передаточной функции нерекурсивного фильтра, соотвутствующего сглаживанию полиномом второй степени по 7 точкам:

Построить графики $\tilde{H}(f)$:

```python
h_7 = lambda x: 1 / 21 * (7 + 12 * np.cos(x) + 6 * np.cos(2 * x) - 4 * np.cos(3 * x))
h_9 = lambda x: 1 / 231 * (59 + 108 * np.cos(x) + 78 * np.cos(2 * x) + 28 * np.cos(3 * x) - 42 * np.cos(4 * x))
h_11 = lambda x: 1 / 429 * (89 + 168 * np.cos(x) + 138 * np.cos(2 * x) + 88 * np.cos(3 * x) + 18 * np.cos(4 * x) - 
                            72 * np.cos( 5 * x))
h_13 = lambda x: 1 / 143 * ( 25 + 48 * np.cos(x) + 42 * np.cos(2 * x) + 32 * np.cos(3 * x) + 18 * np.cos(4 * x) - 
                            22 * np.cos(6 * x))
t = np.linspace(0, math.pi, 64)
plt.plot(t, h_7(t), label=r'$H_7$')
plt.plot(t, h_9(t), label=r'$H_9$')
plt.plot(t, h_11(t), label=r'$H_{11}$')
plt.plot(t, h_13(t), label=r'$H_{13}$')
plt.xlabel('t')
plt.ylabel(r'$H(t)$')
plt.legend()
plt.title("Графики передаточной функции при сглаживании полиномом второй степени по 7, 9, 11 и 13 точкам")
plt.show()
```
