## Основные теоретические положения



В обработке сигналов сглаживание полиномами используется для устранения высокочастотных компонент и шума из сигнала, сохраняя при этом его низкочастотные компоненты и основную информацию. Этот метод сглаживания использует полином для аппроксимации сигнала в каждой точке и удаления высокочастотных колебаний между точками. Сглаживание полиномами может быть реализовано с использованием различных методов, включая метод наименьших квадратов, метод взвешенных наименьших квадратов и другие.

Сглаживание полиномами осуществляется путем применения фильтра к заданному дискретному сигналу. Фильтр представляет собой набор коэффициентов, которые определяют степень сглаживания и влияют на форму сигнала после обработки. В результате применения фильтра сигнал становится более гладким и менее зашумленным.

Спектр дискретного сигнала представляет собой совокупность частот, присутствующих в сигнале. Для определения спектра дискретного сигнала необходимо выполнить его преобразование Фурье.

Преобразование Фурье позволяет преобразовать сигнал из временной области в частотную область и обратно. В результате преобразования получается комплексная функция, модуль которой представляет амплитуду каждой частоты, а аргумент - ее фазу.

Дискретное преобразование Фурье (ДПФ):

$X(n) = \sum\limits_{k=0}^{N-1}y_k e^{-i {{2\pi n} \over N} k}$, $n = 0,1,2,\dots,N-1$

Спектр дискретного сигнала может быть рассчитан с помощью дискретного преобразования Фурье (ДПФ). ДПФ позволяет определить амплитуду и фазу каждой частоты в сигнале, а также оценить ширину спектра и наличие различных гармоник.

1. Сформировать дискретный сигнал посредством дискретизации с шагом $T = 1$ непрерывного сигнала,
представляющего собой линейную комбинацию косинусоид вида $A_k \cos(\omega_k t + \varphi_k)$. Частота каждой из гармоник не
должна превышать $\pi$. Всего одиннадцать гармоник с упорядоченными по возрастанию частотами от О до $\pi$,
изменяющимися с шагом $\Delta \omega = 0.1\pi$. Амплитуды гармоник $A_k$ представляют собой целые числа со значениями от 1
до 11 , определяемые случайным образом с помощью датчика равномерно распределенных случайных чисел. При
необходимости нормализовать коэффициенты линейной комбинации посредством деления их на сумму полученных
случайным образом амплитуд. Начальные фазы $\varphi_k$ представляют собой случайные числа в промежутке от О до 0.5.
Дискретная последовательность должна включать в себя 32 отсчета ($N=31$).

```python
import numpy as np
from scipy.fftpack import fft, ifft, fftshift, rfft, fftfreq
import matplotlib.pyplot as plt
```

```python
def get_analog(n):
    ws = np.arange(0, np.pi + 0.1 * np.pi, 0.1 * np.pi)
    us = np.random.random(size=11)/2.
    As = np.random.randint(1, 11, 11)
    x = np.linspace(0, 32, n)
    y = 0
    for w, u, A in zip(ws, us, As):
        y += A*np.cos(w*x + u)
    y = y / np.sum(As)
    return x, y

def get_discrete(y):
    out_x = np.linspace(0, 32, 32)
    out_y = []
    for idx in range(32):
        out_y.append(y[idx * 8])
    return out_x, np.array(out_y)
```

2. Визуализировать исходные аналоговый и дискретизированный сигналы.

```python
x, y = get_analog(256)
plt.plot(x, y)
plt.ylabel(r'$x$(t)')
plt.xlabel('t')
plt.title('Аналоговый сигнал')
plt.show()

x_disc, y_disc = get_discrete(y)
plt.stem(x_disc, y_disc)
plt.ylabel(r'$x$(t)')
plt.xlabel('t')
plt.title('Дискретный сигнал')
plt.show()
```

3. С помощью ДПФ найти дискретные отсчеты спектра дискретного сигнала и визуализировать их.

Рассмотрим фрагмент последовательности $\{x_k\}$ из отсчётов $N = 20$. Получим дискретную функцию с периодом $NT$, где $T$ - период дискретизации:

$s(t) = \sum\limits_{k = -\infty}^{+\infty} x_k \delta(t - kT)$

Спектр $s(t)$ также должен быть периодическим (с периодом ${{2\pi} \over T}$) и дискретным расстоянием между гармониками ${{2\pi} \over {NT}}$.

Поскольку $s(t)$ - периодическая функция, то мы можем разложить ее в ряд Фурье с помощью ДПФ. Получим дискретные отсчеты спектра и визуализируем их:

```python
x, y = x_disc, y_disc
plt.stem(2 * np.abs(fft(y)) / 32)
plt.ylabel(r'level')
plt.xlabel('freq')
plt.title('Дискретные отсчёты спектра дискретного сигнала')
plt.show()
```

Спектр симметричен относительно нуля, он представляет собой разложение исходного сигнала на линейную комбинацию простых
синусоидальных функций и отражает амплитуды этих функций на разных частотах. Спектр имеет периодичность с шагом 1.

4. Для дискретного сигнала применить линейное сглаживание по 5-ти и 9-ти точкам, представить формулу для $H(\omega)$ – передаточной функции (частотной характеристики) фильтра.

Передаточные функции фильтра для 5 и 9 точек:

$H_5 (\omega) = \frac15 (1 + 2\cos\omega + 2\cos{2\omega})$

$H_9(\omega) = \frac15 (1 + 2\cos\omega + 2\cos{2\omega} + 2\cos{3\omega} + 2\cos{4\omega})$

По данным передаточным функциям выведем формулу для линейного сглаживания по 5 точкам и 9 точкам:

$y_{n,5} = \frac15 \sum\limits_{k=n-2}^{n+2} x_k$

$y_{n,9} = \frac19 \sum\limits_{k=n-4}^{n+4} x_k$

```python
av_y_5 = np.convolve(y, np.ones(5), 'same') / 5
av_y_9 = np.convolve(y, np.ones(9), 'same') / 9
```

5. Визуализировать полученный после фильтрации дискретный сигнал совместно с исходным дискретным сигналом.

```python
plt.stem(x, y)
plt.ylabel(r'$x$(t)')
plt.xlabel('t')
plt.title('Дискретный сигнал')
plt.show()

plt.stem(av_y_5)
plt.ylabel(r'$x$(t)')
plt.xlabel('t')
plt.title('Дискретный сигнал после линейного сглаживания по 5-и точкам')
plt.show()

plt.stem(av_y_9)
plt.ylabel(r'$x$(t)')
plt.xlabel('t')
plt.title('Дискретный сигнал после линейного сглаживания по 9-и точкам')
plt.show()
```

6. С помощью ДПФ найти дискретные отсчеты спектра дискретного сигнала после его фильтрации и визуализировать их совместно с отчетами спектра исходного дискретного сигнала.

Дискретное преобразование Фурье (ДПФ):

$X(n) = \sum\limits_{k=0}^{N-1}y_k e^{-i {{2\pi n} \over N} k}$, $n = 0,1,2,\dots,N-1$

Построим графики:

```python
f = fft(y)
f = 2 * np.abs(f) / len(f)
f_5 = fft(av_y_5)
f_5 = 2 * np.abs(f_5) / len(f_5)
f_9 = fft(av_y_9)
f_9 = 2 * np.abs(f_9) / len(f_9)
plt.stem(f)
plt.ylabel(r'level')
plt.xlabel('freq')
plt.title('Дискретные отсчёты спектра исходного дискретного сигнала')
plt.show()
```

```python
plt.stem(f_5)
plt.ylabel(r'level')
plt.xlabel('freq')
plt.title('Дискретные отсчёты спектра дискретного сигнала после линейного сглаживания по 5-и точкам')
plt.show()
```

```python
plt.stem(f_9)
plt.ylabel(r'level')
plt.xlabel('freq')
plt.title('Дискретные отсчёты спектра дискретного сигнала после линейного сглаживания по 9-и точкам')
plt.show()
```

Из спектра видно, что без ослабления пропускается только сигнал постоянного уровня (нулевой частоты). С увеличением числа точек полоса пропускания становится уже.

7. Проанализировать результат на соответствие значениям $H(\omega)$. Сделать выводы.

Преобразование Фурье сопоставляет непрерывной во времени бесконечной функции $y_n$ другую, непрерывную по частоте, бесконечную функцию - спектр $X(n)$. Графики подтверждают ранее сделанные выводы относительно полосы пропускания и ослабления уровня.
