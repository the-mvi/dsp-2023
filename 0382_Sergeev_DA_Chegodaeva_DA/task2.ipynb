## Основные теоретические положения

* *Дискретный единичный скачок*:
$$\sigma_d(k) = \begin{cases}1, & k\geq0\\ 0,& k < 0\end{cases}, \quad k = ..., -2, -1, 0, 1, 2,...$$

* *Дискретный сигнал* — это сигнал, дискретный по времени и непрерывный по состоянию.

* *Дискретное нормированное время*:
  $$n\in[0, N-1]$$

* *Линейная комбинация* - выражение, равное сумме произведений элементов множества (векторов, функций и т. д.) на числа.

* *Огибающая сигнала* - функция, описывающая изменения какого-либо параметра во времени.

* *Характеристики*:

  *Среднее значение*: $$M = \frac{\sum_{k=0}^{N}s(k)}{N}$$
  *Энергия*: $$E = \sum_{k=0}^{N}{s}^2(k)$$  
  *Средняя мощность*: $$P = \frac{E}{N}$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

$\quad$ Инициализация исходных данных, согласно варианту:

```python
Nb = 14
N = 30 + Nb % 5
a = (-1)**Nb * (0.8 + 0.005 * (Nb%20))
C = 1 + Nb%5
w0 = np.pi/(6 + Nb%5)
U = Nb
n0 = 3 + Nb % 5
n_imp = 5 + Nb % 5
B1 = 1.5 + Nb % 5
B2 = 5.7 - Nb % 5
B3 = 2.2 + Nb % 5
w1 = np.pi/(4 + Nb % 5)
w2 = np.pi/(8 + Nb % 5)
w3 = np.pi/(16 + Nb % 5)
a1 = 1.5 - Nb % 5
a2 = 0.7 + Nb % 5
a3 = 1.4 + Nb % 5

int_n = np.arange(N-1)
```

**1.**  
$\quad$Был смоделирован дискретный прямоугольный импульс $s_1(k)$ на основе дискретного единичного скачка $σ_d(k)$ из прошлой практичекой работы:

```python
def sigma_d(k):
  if k >= 0:
    return 1
  else:
    return 0

def s_1(k):
  return U*(sigma_d(k - n0)-sigma_d(k - n0 - n_imp + 1))
```

$\quad$Построен график значений $s_1$ на интервале времени $n\in [0, N-1]$:

```python
plt.stem(int_n, [s_1(i) for i in int_n], 'purple')
plt.title('Дискретный прямоугольный импульс', color = 'hotpink')
plt.xlabel('n')
plt.ylabel('s_1')

plt.tight_layout()
plt.show()
```

> Моделирование импульса выполнялось на основе представления $s_1$ в виде комбинации двух дискретных единичных скачков $\sigma_d$ с задержкой, учитывающей границы интервала $[n_0 ; (n_0+n_{imp}-1)]$.  Таким образом:
</br>$s_1(k) =  U(\sigma_d(k - n_0)-\sigma_d(k - n_0 - n_{imp} + 1)) $


**2.**

$\quad$Далее была смоделирована линейная комбинация дискретных гармонических сигналов $s_2(k)$:

```python
def x_1(k):
  x1 = B1 * np.sin(w1 * k)
  return x1

def x_2(k):
  x2 = B2 * np.sin(w2 * k)
  return x2

def x_3(k):
  x3 = B3 * np.sin(w3 * k)
  return x3

def s_2(k):
  s2 = a1*x_1(k)+a2*x_2(k)+a3*x_3(k)
  return s2
```

$\quad$Графики последовательностей $x_i(k)$ $i=1,2,3$, и $s_2(k)$ на интервале времени $n\in [0, 5N-1]$:

```python
int_5n = np.arange(5*N-1)

plt.figure(figsize=(14, 10))

plt.subplot(4, 1, 1)
plt.stem(int_5n, [x_1(i) for i in int_5n], 'purple')
plt.title('Дискретный гармоническй сигнал x_1', color = 'hotpink')
plt.xlabel('n')
plt.ylabel('x_1')

plt.subplot(4, 1, 2)
plt.stem(int_5n, [x_2(i) for i in int_5n], 'purple')
plt.title('Дискретный гармоническй сигнал x_2', color = 'hotpink')
plt.xlabel('n')
plt.ylabel('x_2')

plt.subplot(4, 1, 3)
plt.stem(int_5n, [x_3(i) for i in int_5n], 'purple')
plt.title('Дискретный гармоническй сигнал x_3', color = 'hotpink')
plt.xlabel('n')
plt.ylabel('x_3')

plt.subplot(4, 1, 4)
plt.stem(int_5n, [s_2(i) for i in int_5n],'purple')
plt.title('Линейная комбинация дискретных гармонических сигналов s_2', color = 'hotpink')
plt.xlabel('n')
plt.ylabel('s_2')

plt.tight_layout()
plt.show()
```

```python
mean_s2 = np.mean([s_2(i) for i in int_5n])
E = np.sum([s_2(i)**2 for i in int_5n])
P = E/len(int_5n)

print('Среднее значение - mean_s2:', mean_s2)
print('Энергия - E:', E)
print('Средняя мощность - P:', P)
```

> a. Операции при моделировании линейной комбинации сигналов $s_2(k)$:  
Чтобы получить значение $s_2(t)$ в конкретный момент времени $t$ нужно найти значения $x_1(t), x_2(t), x_3(t)$, домножить их на соответствующие коэффициенты и просуммировать.

> b. Определение указанных характеристик:
> *  Среднее значение: $$M = \frac{\sum_{k=0}^{5N-1}s_2(k)}{5N}$$
> *  Энергия: $$E = \sum_{k=0}^{5N-1}{s_2}^2(k)$$  
> *  Средняя мощность: $$P = \frac{E}{5N}$$

**3.**

$\quad$Следом cмоделирован и построен график дискретного сигнала $s_3(k)$, представляющего собой дискретный гармонический сигнал $s(k)$ с экспоненциальной огибающей $|a|^k$ на интервале времени $n\in [0, N-1]$:


```python
def s(k):
  s = C * np.cos(w0 * k)
  return s

def s_3(k):
  s3 = s(k) * np.abs(a)**k
  return s3

plt.figure(figsize=(12, 8))

plt.plot(C*np.abs(a)**int_n, color='hotpink', label='C*|a|^k')
plt.stem(int_n, [s_3(i) for i in int_n], 'purple', label='s_3')
plt.title('Дискретный сигнал s_3', color = 'purple')
plt.xlabel('n')
plt.ylabel('s_3')

plt.legend()
plt.show()
```

> Аналитическая формула сигнала $s_3(k)$ имеет вид: $$s_3(k) = C*cos(\hat{ω_0}k)*|a|^k$$
Данный вид формируется как произведение экспоненциальной огибающей $|a|^k$ и дискретного гармонического сигнала $s(k)$, для того, чтобы результирующий сигнал удовлетворял поставленному условию - огибался $|a|^k$.







**4.**

$\quad$Вывод графика пяти периодов периодической последовательности $s_4(k)$ дискретных прямоугольных импульсов амплитуды $U$ и длительности $n_{imp}$ с периодом, вдвое большим длительности импульса:

```python
period = n_imp*2
interval = np.arange(5*period)

def s_4(k):
  s4=0
  for i in range (0, 5):
    s4 += s_1(k-i*period)
  return s4

plt.figure(figsize=(12, 8))

plt.stem(interval, [s_4(i) for i in interval], 'purple')
plt.title('Периодическая последовательность дискретных прямоугольных импульсов s_4', color = 'hotpink')
plt.xlabel('n')
plt.ylabel('s_4')

plt.show()
```



> Моделирование периодической последовательности $s_4(k)$ основывалось на сумме дискретных прямоугольных импульсов $s_1(k)$ с учётом отступа, соответствующего периоду:
$$s_4(k) = \sum_{i=0}^{N-1}s_1(k-iT)$$ где $T$ - период, $N$ - число периодов
