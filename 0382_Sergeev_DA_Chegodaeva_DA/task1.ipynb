## Основные теоретические положения

* *Единичный цифровой импульс* :
$$\delta_d(k) = \begin{cases}1, & k = 0\\0, & k \neq 0\end{cases}, \quad k = ..., -2, -1, 0, 1, 2,...$$

* *Дискретный единичный скачок*:
$$\sigma_d(k) = \begin{cases}1, & k\geq0\\ 0,& k < 0\end{cases}, \quad k = ..., -2, -1, 0, 1, 2,...$$

* *Дискретная экспоненциальная функция*:
$$s_1(k) = \begin{cases}a^k, & k \ge 0 \\ 0, & k < 0 \end{cases}, \quad k = ..., -2, -1, 0, 1, 2,...$$

* *Дискретный комплексный гармонический сигнал*:
$$s_2(k)=C\exp(j\hatω_0k), \quad k = ..., -2, -1, 0, 1, 2,...$$

* *Аналоговый сигнал* — это сигнал, аргумент функции которого меняется непрерывно.

* *Дискретный сигнал* — это сигнал, дискретный по времени и непрерывный по состоянию.

* *Цифровой сигнал* — это сигнал, дискретный по времени и состоянию.

* *Период дискретизации* — интервал времени, через который берутся отсчеты сигнала.

* *Дискретное время*:
  $$nT\in[0,(N-1)T]$$

* *Дискретное нормированное время*:
  $$n\in[0, N-1]$$

* *Дельта-функция (функция Дирака)*:
$$\delta(k) = \begin{cases}∞, & k = 0\\0, & k \neq 0\end{cases}$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

$\quad$ Инициализация исходных данных, согласно варианту:

```python
Nb = 14
N = 30 + Nb % 5
T = 0.0005 * (1 + Nb%3)
a = (-1)**Nb * (0.8 + 0.005 * (Nb%20))
C = 1 + Nb%5
w0 = np.pi/(6 + Nb%5)
m = 5 + Nb%5

int_n = np.arange(N-1) #norm_discr_time
int_nT = int_n*T #discr_time

print('Nb = %d, N = %d, T = %.4f, a = %.2f, C = %d, w0 = %f, m = %d.' % (Nb, N, T, a, C, w0, m))
```

**1.**  
$\quad$Был смоделирован единичный цифровой импульс $δ_d(k)$:

```python
def delta_d(t):
  if t == 0:
    return 1
  else:
    return 0
```

$\quad$На основе модели постороены графики для цифрового единичного импульса на интервале дискретного и дискретного нормированного времени:

```python
plt.figure(figsize=(14, 7))

plt.subplot(1, 2, 1)
plt.stem(int_n, [delta_d(i) for i in int_n], 'purple')
plt.title('Цифровой единичный импульс на интервале дискретного норм. времени', color = 'purple')
plt.xlabel('n')
plt.ylabel('delta_d')

plt.subplot(1, 2, 2)
plt.stem(int_nT, [delta_d(i) for i in int_nT], 'purple')
plt.title('Цифровой единичный импульс на интервале дискретного времени', color = 'hotpink')
plt.xlabel('nT')
plt.ylabel('delta_d')

plt.tight_layout()
plt.show()
```

> а. Взаимосвязь между дискретным и дискретным нормированным временем основывается на том, что первое строится посредством второго, с учетом  периода дискритизации $T$.

> b. Pазличие между единичным цифровым импульсом и дельта-функцией заключается в том, что значение дельта-функции в точке 0 равно $∞$, а единичного цифрового импульса -- 1.

**2.**

$\quad$Далее был смоделирован дискретный единичный скачок $\sigma_d(k)$:

```python
def sigma_d(t):
  if t >= 0:
    return 1
  else:
    return 0
```

$\quad$Постороение графиков для дискретного единичного скачка на интервалах дискретного и дискретного нормированного времени:

```python
plt.figure(figsize=(14, 7))

plt.subplot(1, 2, 1)
plt.stem(int_n, [sigma_d(i) for i in int_n], 'purple')
plt.title('Дискретный единичный скачок на интервале дискретного норм. времени', color = 'purple')
plt.xlabel('n')
plt.ylabel('sigma_d')

plt.subplot(1, 2, 2)
plt.stem(int_nT, [sigma_d(i) for i in int_nT],'purple')
plt.title('Дискретный единичный скачок на интервале дискретного времени', color = 'hotpink')
plt.xlabel('nT')
plt.ylabel('sigma_d')

plt.tight_layout()
plt.show()
```

> a. Cоответствие между дискретным и аналоговым единичными скачками состоит в том, что аналоговый единичный скачок (в случае, когда значение в точке 0  равняется 1) является частным случаем дискретного единичного скачка с бесконечно малым периодом дискретизации. Следует отдельно рассмотреть точку 0: в случае аналоговой функции значение в данной точки может быть равно 0.5 или неопределено.

> b. Частота дискретизации дискретного единичного скачка:

**3.**

$\quad$Следом была смоделирована дискретная экспоненциальная функция $s_1(k)$:



```python
def s1(t):
  if t >= 0:
    return a**t
  else:
    return 0
```

$\quad$Вывод соотвествующих графиков на интервале дискретного времени и интервале дискретного нормированного времени:

```python
plt.figure(figsize=(14, 7))

plt.subplot(1, 2, 1)
plt.stem(int_n, [s1(i) for i in int_n], 'purple')
plt.title('Дискретная эксп. функция на интервале дискретного норм. времени', color = 'purple')
plt.xlabel('n')
plt.ylabel('s_1')

plt.subplot(1, 2, 2)
plt.stem(int_nT, [s1(i) for i in int_nT], 'purple')
plt.title('Дискретная эксп. функция на интервале дискретного времени', color = 'hotpink')
plt.xlabel('nT')
plt.ylabel('s_1')

plt.tight_layout()
plt.show()
```

>Cоответствие между дискретной и аналоговой экспонентами практически повторяет закономерность между дискретным и аналоговым единичными скачками: <br> $\quad$1) Аналоговая экспонента является частным случаем дискретной экспоненты с бесконечно малым периодом дискретизации; <br> $\quad$2) При значениях аргумента $≤0$ аналоговая экспонента принимает значение $a^k$, а дискретная -- 0.



**4.**

$\quad$Был смоделирован дискретный комплексный гармонический сигнал $s_2(k)=Cexp(j\hat{w_{0}}k)$:

```python
def s2(t):
  return C*np.exp(1j*w0*t)
```

$\quad$Графики вещественной и мнимой части дискретного комплексного гармонического сигнала на интервале дискретного нормированного времени:

```python
plt.figure(figsize=(14, 7))

plt.subplot(1, 2, 1)
plt.stem(int_n, [(s2(i)).real for i in int_n], 'purple')
plt.title('Вещественная часть', color = 'purple')
plt.xlabel('n')
plt.ylabel('s_2')

plt.subplot(1, 2, 2)
plt.stem(int_n, [(s2(i)).imag for i in int_n], 'purple')
plt.title('Мнимая часть', color = 'hotpink')
plt.xlabel('n')
plt.ylabel('s_2')

plt.tight_layout()
plt.show()
```



> Запись данного сигнала в виде комбинации двух вещественных последовательностей:<br> воспользовавшись формулой Эйлера $e^{±iϕ}=cos(ϕ)±isin(ϕ)$ получим, что $$s_2(k)=C\exp(j\hatω_0k) = C(cos(\hatω_0k) + jsin(\hatω_0k))$$

**5.**

$\quad$Построение графиков последовательностей $\delta_d(k)$, $σ_d(k)$ и $s_1(k)$, задержанных на $m=9$ осчётов, на интервале дискретного нормированного времени:

```python
plt.figure(figsize=(21, 7))

plt.subplot(1, 3, 1)
plt.stem(int_n, [delta_d(i-m) for i in int_n], 'purple')
plt.title('Цифровой единичный импульс', color = 'purple')
plt.xlabel('n')
plt.ylabel('delta_d')

plt.subplot(1, 3, 2)
plt.stem(int_n, [sigma_d(i-m) for i in int_n], 'purple')
plt.title('Дискретный единичный скачок', color = 'hotpink')
plt.xlabel('n')
plt.ylabel('sigma_d')

plt.subplot(1, 3, 3)
plt.stem(int_n, [s1(i-m) for i in int_n], 'purple')
plt.title('Дискретная эксп. функция', color = 'purple')
plt.xlabel('n')
plt.ylabel('s_1')

plt.show()
```



> Формулы задержанных последовательностей: <br> *Единичный цифровой импульс:*
$$\delta_d(k-m) = \begin{cases}
 1, & k = m\\0,
 & k \neq m
 \end{cases}$$
*Дискретный единичный скачок:*
$$\sigma_d(k-m) = \begin{cases}
 1, & k \ge m\\0,
 & k < m
 \end{cases}$$
*Дискретная экспоненциальная функция:*
$$s_1(k-m) = \begin{cases}
a^{k-m}, & k \ge m\\0,
 & k < m
 \end{cases}$$
