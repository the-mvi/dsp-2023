#### 1. Вывести формулы передаточных функций рекурсивных фильтров, соответствующих квадратурным формулам прямоугольников, трапеций и Симпсона. Построить графики передаточных функций и графики отношения вычисляемого в результате фильтрации значения к истинному. Проинтерпретировать частотные свойства полученных передаточных функций.

* Формула прямоугольников: $y_{n+1} = y_n + s_{n + \frac 1 2}, y_0 = 0$.<br>
Пусть $s_n = e^{i \omega n}, y_n = H(\omega)e^{i \omega n}$, тогда:<br>
$\begin{cases}
   y_{n+1} = H(\omega)e^{i \omega n} + e^{i \omega (n + 0.5)}\\
   y_{n+1} = H(\omega)e^{i \omega (n + 1)}
 \end{cases}$<br>

 $H(\omega)e^{i \omega (n + 1)} = H(\omega)e^{i \omega n} + e^{i \omega (n + 0.5)}$<br>

 $H(\omega)(e^{i \omega n}e^{i \omega}) = H(\omega)e^{i \omega n} + e^{i \omega n}e^{0.5 i \omega }$<br>

 $H(\omega)(e^{i \omega n}e^{i \omega}-e^{i \omega n}) = e^{i \omega n}e^{0.5 i \omega }$<br>

 $H(\omega)(e^{i \omega} - 1) = e^{0.5 i \omega }$<br>

 $H(\omega) = \frac {e^{0.5 i \omega }} {e^{i \omega} - 1} = \frac {1} {e^{\frac 1 2 i \omega} - e^{- \frac 1 2 i \omega}} = \frac 1 {2i \sin {\frac \omega t}}$<br>

 $\tilde H(f) = \frac 1 {2i \sin {\pi f}}$<br>

 Отношение значений:<br>
  $\gamma = \frac {\text {Вычисленное}}{\text {Точное}} = \frac {\frac \omega 2}{\sin {\frac \omega 2}}$


* Формула трапеций: $y_{n+1} = y_n + \frac 1 2(s_n + s_{n+1}), y_0 = 0$.<br>

 Пусть $s_n = e^{i \omega n}, y_n = H(\omega)e^{i \omega n}$, тогда:<br>

 $\begin{cases}
   y_{n+1} = H(\omega)e^{i \omega n} + \frac 1 2 (e^{i \omega n} + e^{i \omega (n + 1)})\\
   y_{n+1} = H(\omega)e^{i \omega (n + 1)}
 \end{cases}$<br>

 $H(\omega)e^{i \omega (n + 1)} = H(\omega)e^{i \omega n} + \frac 1 2 (e^{i \omega n} + e^{i \omega (n + 1)})$<br>

 $H(\omega)(e^{i \omega n}e^{i \omega}) = H(\omega)e^{i \omega n} + \frac 1 2 (e^{i \omega n} + e^{i \omega n}e^{i \omega})$<br>

 $H(\omega)(e^{i \omega}) = H(\omega) + \frac 1 2 (1 + e^{i \omega})$<br>

 $H(\omega)(e^{i \omega} - 1) = \frac 1 2 (1 + e^{i \omega})$<br>

 $H(\omega) = \frac {e^{i \omega} + 1} {2(e^{i \omega} - 1)} = \frac {\cos {\frac \omega 2}}{2i \sin {\frac \omega 2}}$<br>

 $\tilde H(f) = \frac {\cos {\pi f}} {2i \sin {\pi f}}$<br>

  Отношение значений:

  $\gamma = \frac {\text {Вычисленное}}{\text {Точное}} = \cos (\frac \omega 2) \frac {\frac \omega 2}{\sin (\frac \omega 2)} = 1 - \frac {\omega^2}{12} + \frac {\omega^4}{720} + \cdots$<br>


* Формула Симпсона: $y_{n+1} = y_{n-1} + \frac 1 3 (s_{n-1} + 4s_n + s_{n+1}),y_0 = 0$<br>

 Пусть $s_n = e^{i \omega n}, y_n = H(\omega)e^{i \omega n}$, тогда:<br>

 $\begin{cases}
   y_{n+1} = H(\omega)e^{i \omega (n-1)} + \frac 1 3 (e^{i \omega (n-1)} + 4e^{i \omega n} + e^{i \omega (n + 1)})\\
   y_{n+1} = H(\omega)e^{i \omega (n + 1)}
 \end{cases}$<br>

 $H(\omega)(e^{i \omega n}e^{i \omega}) = H(\omega)(e^{i \omega n}e^{-i \omega}) + \frac 1 3 e^{i \omega n} (e^{-i \omega} + 4 + e^{i \omega})$<br>

 $H(\omega)e^{i \omega} = H(\omega)e^{-i \omega} + \frac 1 3(e^{-i \omega} + 4 + e^{i \omega})$<br>

 $H(\omega)(e^{i \omega} - e^{-i \omega}) = \frac 1 3(e^{-i \omega} + 4 + e^{i \omega})$<br>

 $H(\omega) = \frac {e^{-i \omega} + 4 + e^{i \omega}} {3(e^{i \omega} - e^{-i \omega})} = \frac {\cos {\omega} + 2}{3i \sin {\omega}}$<br>

  $\tilde H(f) = \frac {\cos {2\pi f} + 2} {3i \sin {2 \pi f}}$<br>

  Отношение значений:

  $\gamma = \frac {\text {Вычисленное}}{\text {Точное}} =  \frac {\cos {\omega} + 2} {3} \cdot \frac {\omega}{\sin {\omega}} = 1 + \frac {\omega^4}{180} + \cdots$<br>

```python
def tilde_H_rect(x):
  return (1 / (2j * np.sin(math.pi * x))).imag
def tilde_H_trap(x):
  return (np.cos(math.pi * x) / (2j * np.sin(math.pi * x))).imag
def tilde_H_simp(x):
  return ((np.cos(2 * math.pi * x) + 2) / (3j * np.sin(2 * math.pi * x))).imag

def y_rect(x):
  return math.pi * x / np.sin(math.pi * x)
def y_trap(x):
  return np.cos(math.pi * x) * (math.pi * x) / (np.sin(math.pi * x))
def y_simp(x):
  return (np.cos(2 * math.pi * x) + 2) / 3 * (2 * math.pi * x) / (np.sin(2 * math.pi * x))


x = np.linspace(1e-10, .5, 300)
```

```python
plt.figure(figsize=(9, 7))
plt.plot(x, tilde_H_rect(x), label=r'Квадратурная формула прямоугольников')
plt.plot(x, tilde_H_trap(x), label=r'Квадратурная формула трапеций')
plt.plot(x, tilde_H_simp(x), label=r'Квадратурная формула Симпсона')
plt.xlabel('f')
plt.ylabel(r'$\tilde{H}(f)$')
plt.legend()
x1, x2, y1, y2 = plt.axis()
plt.axis((x1, x2, -5, 1.))
plt.title(r'Передаточная функция $\tilde {H} (f)$ рекурсивного фильтра')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(9, 7))
plt.plot(x, y_rect(x), label=r'$\gamma_{rect}$')
plt.plot(x, y_trap(x), label=r'$\gamma_{trap}$')
plt.plot(x, y_simp(x), label=r'$\gamma_{simp}$')
plt.xlabel('f')
plt.ylabel(r'$\gamma(f)$')
plt.legend()
x1, x2, y1, y2 = plt.axis()
plt.axis((x1, x2, -0.1, 2.))
plt.title(r'Отношения вычисляемого к истинному')
plt.grid()
plt.show()
```

Из графиков видно, что рекурсивный фильтр, представленный формулой трапеции, эффективно подавляет высокие частоты. В отличие от этого, квадратурные формулы прямоугольников и Симпсона на графиках проявляют склонность к усилению высоких частот.

#### 2. Вывести формулу передаточной функции рекурсивного фильтра для интегрирования «по правилу 3/8»: $y_{n+2} = y_{n-1} + \frac 1 8 (s_{n+2} + 3s_{n+1} + 3s_n + s_{n-1})$.
Построить график передаточной функции и график отношения вычисляемого в результате фильтрации значения к истинному. Проинтерпретировать частотные свойства передаточной функции.

Квадратурная формула «по правилу 3/8»: $y_{n+2} = y_{n-1} + \frac 1 8 (s_{n+2} + 3s_{n+1} + 3s_n + s_{n-1})$.

Пусть $s_n = e^{i \omega n}, y_n = H(\omega)e^{i \omega n}$, тогда:<br>

 $\begin{cases}
   y_{n+2} = H(\omega)e^{i \omega (n-1)} + \frac 1 8 (e^{i \omega (n+2)} + 3e^{i \omega (n+1)} + 3e^{i \omega n} + e^{i \omega (n - 1)})\\
   y_{n+2} = H(\omega)e^{i \omega (n + 2)}
 \end{cases}$<br>

 $H(\omega)e^{i \omega (n + 2)} = H(\omega)e^{i \omega (n-1)} + \frac 1 8 (e^{i \omega (n+2)} + 3e^{i \omega (n+1)} + 3e^{i \omega n} + e^{i \omega (n - 1)})$<br>

 $H(\omega) (e^{2 i \omega} - e^{- i \omega}) = \frac 1 8 (e^{2i \omega} + 3e^{i \omega} + 3 + e^{- i \omega})$<br>

 $H(\omega) = \frac {e^{2i \omega} + 3e^{i \omega} + 3 + e^{- i \omega}}{8 (e^{2 i \omega} - e^{- i \omega})} \cdot \frac {e^{- \frac {i\omega}{2}}}{e^{- \frac {i\omega}{2}}} = \frac {e^{\frac {3i\omega}{2}} + 3e^{\frac {i\omega}{2}} + 3e^{- \frac {i\omega}{2}} + e^{- \frac {3i\omega}{2}}}{8 (e^{\frac {3i\omega}{2}} -\ e^{- \frac {3i\omega}{2}})} = \frac {2 \cos {(\frac {3\omega} 2)} + 6 \cos{(\frac \omega 2)}}{16 i \sin (\frac {3\omega} 2)}$<br>

 $\tilde H (\omega) = \frac {\cos {(3 \pi f)} + 3 \cos{(\pi f)}}{8 i \sin (3 \pi f)}$<br>

 Отношение значений:
 $\gamma = \frac {\text {Вычисленное}}{\text {Точное}} =  \frac {(2 \cos {(\frac {3\omega} 2)} + 6 \cos{(\frac \omega 2)}) i\omega}{16 i \sin (\frac {3\omega} 2)} = \frac {(\cos {(\frac {3\omega} 2)} + 3 \cos{(\frac \omega 2)}) \omega}{8 \sin (\frac {3\omega} 2)} = \frac 1 {12} (\cos {(\frac {3\omega} 2)} + 3 \cos{(\frac \omega 2)}) \cdot \frac {\frac {3\omega}{2}}{\sin (\frac {3\omega}{2})} = \frac 1 3 + \frac {\omega^4}{240} + \cdots$<br>

 $\gamma = \frac 1 {12} (\cos {(3 \pi f)} + 3 \cos{(\pi f)}) \cdot \frac {3 \pi f}{\sin (3 \pi f)} = \frac 1 3 + \frac {\pi^4 f^4}{15} + \cdots$<br>

```python
def tilde_H(x):
  return ((np.cos(3 * math.pi * x) + 3 * np.cos(math.pi * x)) / (8j * np.sin(3 * math.pi * x))).imag

def y(x):
  return 1/12 * (np.cos(3 * math.pi * x) + 3 * np.cos(math.pi * x)) * (3 * math.pi * x) / (np.sin(3 * math.pi * x))

x = np.linspace(1e-10, .5, 300)
```

```python
plt.plot(x, tilde_H(x), label=r'правило 3/8')
plt.xlabel('f')
plt.ylabel(r'$\tilde{H}(f)$')
plt.legend()
x1, x2, y1, y2 = plt.axis()
plt.axis((x1, x2, -50, 10.))
plt.title(r'Передаточная функция $\tilde {H} (f)$ рекурсивного фильтра')
plt.grid()
plt.show()
```

```python
plt.plot(x, y(x), label=r'$\gamma$')
plt.xlabel('f')
plt.ylabel(r'$\gamma(f)$')
plt.legend()
x1, x2, y1, y2 = plt.axis()
plt.axis((x1, x2, -15., 27.))
plt.title(r'Отношение вычисляемого в результате фильтрации значения к истинному')
plt.grid()
plt.show()
```

По результатам графиков видно, что на частоте ~0,33Гц происходят скачки: сначала сигнал резко увеличивается, а затем стремительно уменьшается. На остальной области частот значения близки к 0.
