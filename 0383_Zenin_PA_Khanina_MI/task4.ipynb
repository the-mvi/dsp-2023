```python
#Импорт библиотек и задание переменных в соответствии с условием

import matplotlib.pyplot as plt
import numpy as np
from math import pi
```

#### **1. Вывести формулы передаточных функций рекурсивных фильтров, соответствующих квадратурным формулам прямоугольников, трапеций и Симпсона. Построить графики передаточных функций и графики отношения вычисляемого в результате фильтрации значения к истинному. Проинтерпретировать частотные свойства полученных передаточных функций:**

Пусть $s_n =  e^{i\omega n}, \text{  } y_n = H(\omega)e^{i\omega n}$, следовательно, для формулы **трапеций**:

$y_{n+1} = y_n + \frac{1}{2}(s_n + s_{n+1})$


$\begin{cases}
y_{n+1} = H(\omega)e^{i\omega n} + \frac{1}{2}(e^{i\omega n} + e^{i \omega (n+1)})\\
y_{n+1} = H(\omega)e^{i\omega (n+1)}\\
\end{cases}$

$H(\omega)e^{i\omega(n+1)} = H(\omega)e^{i\omega n} + \frac{1}{2}(e^{i\omega n} + e^{i\omega(n+1)})$

$H(\omega)(e^{i\omega}) = H(\omega) + \frac{1}{2}(1+e^{i\omega})$

$H(\omega)(e^{i\omega}-1) = \frac{1}{2}(1 + e^{i\omega})$

$H(\omega) = \frac{cos{\frac{\omega}{2}}}{2i \sin{\pi f}}$

$\tilde{H}(f) = \frac{\cos\pi f}{2i\sin \pi f}$

$\gamma = \frac{\text{Вычисленное}}{\text{Точное}} = \cos(\frac{\omega}{2})\frac{\frac{\omega}{2}}{\sin(\frac{\omega}{2})}$

$\tilde{\gamma} = \frac{\text{Вычисленное}}{\text{Точное}} = \cos(\pi f)\cdot\frac{\pi f}{\sin(\pi f)}$

_____________________

Для формулы **прямоугольников**:

$y_{n+1} = y_n + s_{n+\frac{1}{2}}, y_0 = 0$

$\begin{cases}
y_{n+1} = H(\omega)e^{i\omega n} + e^{i \omega (n + \frac{1}{2})}\\
y_{n+1} = H(\omega)e^{i \omega (n+1)}
\end{cases}$

$H(\omega)e^{i \omega (n+1)} = H(\omega)e^{i\omega n} + e^{i \omega (n + \frac{1}{2})}$

$H(\omega)(e^{i \omega n}\cdot e^{i\omega} - e^{i\omega n}) = e^{i\omega n} \cdot e^{\frac{1}{2}i\omega}$

$H(\omega)(e^{i\omega} - 1) = e^{\frac{1}{2}i\omega}$

$H(\omega) = \frac{1}{2i\sin(\frac{\omega}{2})}$

$\tilde{H}(f) = \frac{1}{2i\sin\pi f}$


$\gamma = \frac{\text{Вычисленное}}{\text{Точное}} = \frac{\frac{\omega}{2}}{\sin\frac{\omega}{2}}$

$\tilde{\gamma} = \frac{\text{Вычисленное}}{\text{Точное}} = \frac{\pi f}{\sin(\pi f)}$

___________

Для формулы **Симпсона**:

$y_{n+1} = y_{n-1} + \frac{1}{3}(s_{n-1} + 4s_n + s_{n+1}), y_0 = 0$


$\begin{cases}
y_{n+1} = H(\omega)e^{i\omega (n-1)} + \frac{1}{3}(e^{i\omega(n-1)} + 4e^{i\omega n} + e^{i\omega (n+1)})\\
y_{n+1} = H(\omega)e^{i\omega(n+1)}
\end{cases}$

$H(\omega)e^{i\omega(n+1)} = H(\omega)e^{i\omega (n-1)} + \frac{1}{3}(e^{i\omega(n-1)} + 4e^{i\omega n} + e^{i\omega (n+1)})$

$H(\omega)(e^{i\omega - e^{-i\omega}}) = \frac{1}{3}(e^{-i\omega} + 4 + e^{i\omega})$

$H(\omega) = \frac{\cos (\omega) + 2}{3i\sin \omega}$

$\tilde{H}(f) = \frac{\cos 2\pi f + 2}{3i \sin 2\pi f}$


$\gamma = \frac{\text{Вычисленное}}{\text{Точное}} = \frac{\cos(\omega)+2}{3} \cdot \frac{\omega}{\sin(\omega)}$

$\tilde{\gamma} = \frac{\text{Вычисленное}}{\text{Точное}} = \frac{\cos(2\pi f)+2}{3} \cdot \frac{2\pi f}{\sin(2\pi f)}$


```python
x = np.linspace(1e-3,0.5,1000)

H_tr = lambda x: np.imag((np.cos(pi * x)/(2j * np.sin(pi * x))))   #п.ф. ф-лы трапеции
H_rc = lambda x: np.imag((1/(2j * np.sin(pi*x))))                  #п.ф. ф-лы прямоугольника
H_Sm = lambda x: np.imag(((np.cos(x) + 2)/(3j * np.sin(2*pi*x))))  #п.ф. ф-лы Симпсона

g_tr = lambda x: (pi*x) / (np.sin(pi*x)) 
g_rc = lambda x: (np.cos(pi*x))*((pi*x) / (np.sin(pi*x)))
g_Sm = lambda x: ((np.cos(2*pi*x)+2)/3)*((2*pi*x)/(np.sin(2*pi*x)))


plt.figure(figsize=(12,12))

plt.title("График передаточных функций")
plt.plot(x, H_tr(x), label="Формула трапеций")
plt.plot(x, H_rc(x), label="Формула прямоугольников")
plt.plot(x, H_Sm(x), label="Формула Симпсона")

plt.xlabel("f")
plt.ylabel(r"$\tilde{H}(f)$")
plt.ylim(-10,3)
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12,12))

plt.title("График отношений вычисленных значений к точным ")
plt.plot(x, g_tr(x), label="Формула трапеций")
plt.plot(x, g_rc(x), label="Формула прямоугольников")
plt.plot(x, g_Sm(x), label="Формула Симпсона")

plt.xlabel("f")
plt.ylabel(r"$\gamma(f)$")
plt.legend()
plt.ylim(0,2)
plt.grid()
plt.show()
```

На графике передаточных функций можно заметить, что формула трапеций даёт уменьшение амплитуды высокочастотных сигналов, тогда как формулы Симпсона и прямоугольника дают усиление высокочастотных сигналов.

На части спектра до $f \approx 0.33$ лучшее отношение вычисленного значения к истинному, среди представленных, демонстрирует формула Симпсона.

#### 2. Вывести формулу передаточной функции рекурсивного фильтра для интегрирования «по правилу 3/8». Построить график передаточной функции и график отношения вычисляемого в результате фильтрации значения к истинному. Проинтерпретировать частотные свойства передаточной функции.


$y_{n+2} = y_{n-1} + \frac{1}{8}(x_{n+2} + 3x_{n+1} + 3x_n + x_{n-1}), y_0 = 0$

$\begin{cases}
y_{n+2} = H(\omega)e^{i \omega (n + 2)}\\
y_{n+2} = H(\omega)e^{i \omega (n-1)} + \frac 1 8 (e^{i \omega (n+2)} + 3e^{i \omega (n+1)} + 3e^{i \omega n} + e^{i \omega (n - 1)})
\end{cases}
$

$H(\omega)e^{i \omega (n + 2)} = H(\omega)e^{i \omega (n-1)} + \frac 1 8 (e^{i \omega (n+2)} + 3e^{i \omega (n+1)} + 3e^{i \omega n} + e^{i \omega (n - 1)})$

$H(\omega)(e^{2 i \omega} - e^{-i \omega}) = \frac 1 8 (e^{2 i \omega } + 3e^{ i \omega} + 3 + e^{-i \omega})
\Rightarrow 
\tilde{H}(f) = \frac{\cos(3\pi f) + 3\cos(\pi f)}{8i\sin(3\pi f)}$ 



$\gamma = \frac{\text{Вычисленное}}{\text{Точное}} = \frac{\cos(\frac{3\omega}{2}) + 3\cos(\frac{\omega}{2})}{12} \cdot \frac{3\omega}{2\sin(\frac{3\omega}{2})}$

$\tilde{\gamma} = \frac{\text{Вычисленное}}{\text{Точное}} = \frac{\cos(3\pi f) + 3\cos(\pi f)}{12} \cdot \frac{3\pi f}{\sin(3\pi f)}$

```python
H_te = lambda x: np.imag(((np.cos(3*pi*x) + 3*np.cos(pi*x)) / (8j * np.sin(3*pi*x))))
g_te = lambda x: ((np.cos(3*pi*x) + 3*np.cos(pi*x))/12) * ((3*pi*x)/(np.sin(3*pi*x)))

plt.figure(figsize=(12,12))

plt.title("График передаточной функций")
plt.plot(x, H_te(x), label="Формула 3/8")

plt.xlabel("f")
plt.ylabel(r"$\tilde{H}(f)$")
plt.ylim(-20,20)
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12,12))

plt.title("График отношений вычисленных значений к точным ")
plt.plot(x, g_te(x), label="Формула 3/8")

plt.xlabel("f")
plt.ylabel(r"$\gamma(f)$")
plt.legend()
plt.grid()
plt.show()
```

На графке видны следующие частотные свойства:

По большей части формула правила 3/8 подавляет низкие частоты до $f=1/3$, затем происходит резкий скачок передаточной фукнции, в ходе которого происходит резкое усиление всех частот вблизи $f=1/3$. По мере приближения к $f=0.5$ амлплитуда частот падает до нуля.
