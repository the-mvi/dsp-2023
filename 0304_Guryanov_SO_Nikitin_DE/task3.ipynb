## Основные теоретические положения

Вывод формулы полиномиального сглаживания первой степени временных рядов:  

$$F(A,B)=\sum^{2}_{k=-2}(s_k-y_k)^2=\sum^{2}_{k=-2}(s_k-A-Bk)^2 \Rightarrow min$$

Система нормальных уравненией:

$$\begin{cases}5A+0B=\sum^{k=2}_{k=-2}s_k\\0A+10B=\sum^{k=2}_{k=-2}ks_k\end{cases}$$

Решение системы:
$$y_0=A=\frac{1}{5}\sum_{k=-2}^{k=2}s_k=\frac{1}{5}(s_{-2}+s_{-1}+s_0+s_1+s_2)$$

$$y_n=\frac{1}{5}\sum_{k=n-2}^{k=n+2}s_k=\frac{1}{5}(s_{n-2}+s_{n-1}+s_n+s_{n+1}+s_{n+2})$$

$$y_n=\frac{1}{5}(e^{-2i\omega}+e^{-i\omega}+1+e^{i\omega}+e^{2i\omega})e^{i\omega n}=H(\omega)e^{i\omega n}$$

Тогда передаточная функция:

$$H(\omega)=\frac{1}{5}(e^{-2i\omega}+e^{-i\omega}+1+e^{i\omega}+e^{2i\omega})$$


$$H(\omega)=0.2[1+2cos(\omega)+2cos(2\omega)]$$

Сумма геометрической прогрессии:

$$H(\omega)=\frac{e^{i\frac{5\omega}{2}}-e^{-i\frac{5\omega}{2}}}{5(e^{i\frac{\omega}{2}}-e^{-i\frac{\omega}{2}})}=\frac{sin(\frac{5\omega}{2})}{5sin(\frac{\omega}{2})}$$

$H(\omega)$ - периодическая функция с периодом $2\pi$. Частота рассматриваемого дискретного сигнала не должна превышать $\pi$, поэтому обычно рассматривается интервал $(-\pi, \pi)$ для круговой частоты $\omega$, или $(-0.5,0.5)$ для циклической частоты $f$:

$$H(\omega)=H(2\pi f)=\tilde H(f)$$

В общем случае при приближении полиномом нулевого порядка прямой линией для $2m+1$ точек:

$$H(\omega)=\frac{1}{2m+1}[1+2cos(\omega)+2cos(2\omega)+...+2cos(m\omega)]$$

$$H(\omega)=\frac{sin(\frac{(2m+1)\omega}{2})}{(2m+1)sin(\frac{\omega}{2})}$$  

При сглаживании полиномом второй степени $y(t)=A+Bt+Ct^2$:

$$F(A,B,C)=\sum^{3}_{k=-3}(s_k-y_k)^2=\sum^{3}_{k=-3}(s_k-A-Bk-Ck^2)^2 \Rightarrow min$$

## Выполнение работы

#### 1. Вывести формулы для передаточной функции нерекурсивного фильтра, соответствующего сглаживанию прямой линией по 3, 5, 7 и 9 точкам. Построить графики $\tilde{H}(f)$. Проинтерпретировать частотные свойства передаточных функций для различного количества точек.

```python
import numpy as np
from matplotlib.pyplot import *
from matplotlib import pyplot
from math import pi

def h(f, n):
    return np.sin(n*pi*f) / (n*np.sin(pi*f))


f = np.linspace(-0.5, 0.5, 100)
for n in [3, 5, 7, 9]:
    plot(f, h(f, n), label=fr'$\tilde H_{n}(f)$')
xlabel('f')
pyplot.ylabel(r'$\tilde H(f)$')
pyplot.legend()
pyplot.show()
```

Количество экстремумов передаточной функции равно количеству точек, использовавшемуся при построении. Наибольшее значение, равное 1, достигается в 0. Функция убывает по модулю в окрестности точки 0. Область, в которой она убывает, тем меньше, чем больше использовалось точек.

#### 2. Вывести формулы для передаточной функции нерекурсивного фильтра, соответствующего сглаживанию полиномом второй степени по 7, 9, 11 и 13 точкам. Построить графики $\tilde{H}(f)$. Проинтерпретировать частотные свойства передаточных функций для различного количества точек.

```python
def h(f, n):
    match n:
        case 7:
            return 1 / 21 * (7 + 12 * np.cos(2*pi*f) + 6 * np.cos(4*pi*f) - 4 * np.cos(6*pi*f))
        case 9:
            return 1 / 231 * (59 + 108 * np.cos(2*pi*f) + 78 * np.cos(4*pi*f) + 28 * np.cos(6*pi*f) - 42 * np.cos(8*pi*f))
        case 11:
            return 1 / 429 * (89 + 168 * np.cos(2*pi*f) + 138 * np.cos(4*pi*f) + 88 * np.cos(6*pi*f) + 18 * np.cos(8*pi*f) - 72 * np.cos(10*pi*f))
        case 13:
            return 1 / 143 * (25 + 48 * np.cos(2*pi*f) + 42 * np.cos(4*pi*f) + 32 * np.cos(6*pi*f) + 18 * np.cos(8*pi*f) - 22 * np.cos(12*pi*f))
for n in [7, 9, 11, 13]:
    plot(f, h(f, n), label=fr'$\tilde H_{{{n}}}(f)$')
xlabel('f')
pyplot.ylabel(r'$\tilde H(f)$')
pyplot.legend()
pyplot.show()
```

Количество экстремумов передаточной функции равно количеству точек - 2, использовавшемуся при построении. Наибольшее значение, равное 1, достигается в 0. Функция убывает по модулю в окрестности точки 0. Область, в которой она убывает, тем меньше, чем больше использовалось точек.
