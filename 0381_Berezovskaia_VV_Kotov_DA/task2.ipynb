```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
## Входные данные
def x1(k):
    return variables['B1'] * np.sin(variables['w1'] * k)

def x2(k):
    return variables['B2'] * np.sin(variables['w2'] * k)

def x3(k):
    return variables['B3'] * np.sin(variables['w3'] * k)

variables = {'Nb': 9}
variables['N'] = 30 + variables['Nb'] % 5
variables['T'] = 0.0005 * (1 + variables['Nb'] % 3)
variables['a'] = (-1) ** variables['Nb'] * (0.8 + 0.005 * variables['Nb'])
variables['C'] = 1 + variables['Nb'] % 5
variables['w0'] = math.pi / (6 + variables['Nb'] % 5)
variables['U'] = variables['Nb']
variables['n0'] = variables['Nb'] % 5 + 3
variables['n_imp'] = variables['Nb'] % 5 + 5
variables['B1'] = 1.5 + variables['Nb'] % 5
variables['B2'] = 5.7 - variables['Nb'] % 5
variables['B3'] = 2.2 + variables['Nb'] % 5
variables['w1'] = math.pi / (4 + variables['Nb'] % 5)
variables['w2'] = math.pi / (8 + variables['Nb'] % 5)
variables['w3'] = math.pi / (16 + variables['Nb'] % 5)
variables['a1'] = 1.5 - variables['Nb'] % 5
variables['a2'] = 0.7 + variables['Nb'] % 5
variables['a3'] = 1.4 + variables['Nb'] % 5


variables['x1'] = x1
variables['x2'] = x2
variables['x3'] = x3

x = np.linspace(0, (variables['N'] - 1) * variables['T'])
x_norm = np.linspace(0, variables['N'] - 1, 33)

def rect(x, m=variables['n0']):
    y = np.zeros(x.shape)
    condition = (m <= x) & (x <= m + variables['n_imp'] - 1)
    y[condition] = variables['U']
    return y
```

### 1. Дискретный прямоугольный импульс:

$s_1(k) = \begin{equation*}
\begin{cases}
   U, n_0 \leq k \leq n_0 + n_{imp} - 1\\
   0, иначе
 \end{cases}
\end{equation*}$

на основе дискретного единичного скачка $\sigma_d(k)$ из л/р №1 с выводом графика на интервале времени $n \in [0,N-1]$.

```python
plt.stem(x_norm, rect(x_norm))
plt.grid(True)
plt.show()
```

Прямоугольный импульс можно представить как линейную комбинацию двух задержанных дискретных единичных скачков $\sigma_d(k)$
 с амплитудами равными по модулю и разными по знаку:

 $s_1(k)=U(σ_d(k-n_0)-σ_d(k-n_0-n_{imp}))$



Для моделирование прямоугольного импульса выполняется 2 шага:
1. Генерируется массив из нулей. Количество нулей равно количетву интервалов
2. Каждому отсчету, удовлетворяющему условию, присваивается $U$.

### 2. Моделирование линейной комбинации дискретных гармонических сигналов $s_2(k)$:

$s_2(k) = a_1x_1(k) + a_2 x_2 (k) + a_3 x_3 (k)$, где $x_i(k) = B_i\sin(\hat\omega_i k)$, $i = 1,2,3$

с выводом графиков последовательностей $x_i(k)$ и $s_2(k)$ на интервале времени $n \in [0, 5N - 1]$ и вычислим среднее значение (идентификатор `mean_s2`), энергию (идентификатор `Е`) и среднюю мощность последовательности $s_2(k)$ (идентификатор `Р`)

```python
def s2(x, variables):
    return (
        variables['a1'] * variables['x1'](x) +
        variables['a2'] * variables['x2'](x) +
        variables['a3'] * variables['x3'](x)
    )

x_4 = np.linspace(0, 5 * variables['N'] - 1, 5 * variables['N'])
s2_sequence = s2(x_4, variables)
```

```python

fig, axes = plt.subplots(2, 2, figsize=(10, 8))
fig.suptitle("Графики последовательностей")


for i, (sequence, title) in enumerate([
    (s2_sequence, "График последовательности s2(k)"),
    (variables['x1'](x_4), "График последовательности x1(k)"),
    (variables['x2'](x_4), "График последовательности x2(k)"),
    (variables['x3'](x_4), "График последовательности x3(k)")
]):
    row, col = i // 2, i % 2
    axes[row, col].stem(x_4, sequence)
    axes[row, col].set_title(title)
    axes[row, col].grid(True)


plt.tight_layout()
plt.show()
```

```python
print('Среднее(mean_s2) - {}, Энергия(E) - {}, Мощность(P) - {}'.format(s2_sequence.mean(), np.power(s2_sequence, 2).sum(), np.power(s2_sequence, 2).mean()))
```

Операции при моделировании линейной комбинации сигналов:

1. Производится вычисление дискретного нормированного времени
2. Вычисляются дискретные гармоники
3. Считается линейная комбинация дискретных гармоник

Определение характеристик:

1. Среднее значение: $M = {1 \over N} \sum\limits_{i=0}^{N-1} x_i=0.7892$

2. Энергия: $E = \sum\limits_{i=0}^{N-1} x^2=110847.7892$

3. Мощность: $P = {{\sum\limits_{i=0}^{N-1} x^2} \over N}=652.0458$

### 3. График дискретного сигнала $s_3(k)$, представляющего собой дискретный гармонический сигнал $s(k)$: $s(k) = C\cos(\hat\omega_0 k)$ с экспоненциальной огибающей $|a|^k$, на интервале времени $n \in [0, N-1]$.

```python
def s3(x, variables):
    return (np.abs(variables['a']) ** x) * np.cos(variables['w0'] * x) * variables['C']

s3_sequence = s3(x_norm, variables)

plt.stem(x_norm, s3_sequence)
plt.grid(True)
plt.title("График дискретной затухающей синусоиды s3(k)")

plt.show()
```

Операции при моделировании данного сигнала:

1. Расчёт времени на интервале $n \in [0, N-1]$.
2. Расчёт дискретного гармонического сигнала.

Аналитическая формула дискретного сигнала $s_3$:

$s_3(k) = |a|^k C \cos(\hat\omega_0 k)$

### 4. График пяти периодов периодической последовательности $s_4(k)$ дискретных прямоугольных импульсов амплитуды $U$ и длительности $n_{imp}$ с периодом, вдвое большим длительности импульса.

```python
x_5 = np.linspace(0, 90, 3 * variables['N'])
y = np.zeros(shape=x_5.shape)

for i in range(5):
    y += rect(x_5, 2 * variables['n_imp'] * i)

plt.stem(x_5, y)
plt.grid(True)
plt.title("График пяти периодов периодической последовательности s4(k)")
plt.show()
```
