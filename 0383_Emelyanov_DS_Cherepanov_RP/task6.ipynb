## Основные теоретические положения
Пусть $\{x_k\}$ — последовательность дискретных отсчетов с периодом $N: \forall k x_k = x_{k+N}$.

Рассмотрим последовательность $\{x_0, \dots, x_{N-1}\}$.

Спектр $s(t)$ должен быть периодическим. Найдем его коэффициенты:
$$X(n) = \frac{1}{NT}\int_0^{NT}s(t)e^{-i\omega_n t}dt.$$
Дискретный аналог формулы без множителя — **дискретное преобразование Фурье**:
$$X(n) = \sum_0^{N-1}x_ke^{-i\frac{2\pi n}{N}k}dt.$$

Пусть $\{x_k\}, \{y_k\}$ — дискретные последовательности, $\{X(n)\}, \{Y(n)\}$ — соответствующие ДПФ. Тогда ДПФ обладает следующими свойствами:
* **Линейность**: ДПФ$[\alpha\{x_k\} + \beta\{y_k\}] = \alpha \{X(n)\} + \beta \{Y(n)\}$;
* **Задержка**: $\{y(k)\} = \{x(k-1)\} \implies \{Y(n)\} = \{X(n)\exp\frac{-2i\pi n}{N}\}$;
* **Симметрия**: $X(N - n) = X(-n) = X^*(n)$ — имеет место для вещественного сигнала;

$H(\omega) = H(2\pi f) = \tilde H(f).$

**Формула численного дифференцирования**: $$s\prime = \frac{s_{n+1} - s_{n-1}}{2}.$$

Рассмотрим последовательность $s_n = e^{i\omega n}.$

Найдем значение передаточной функции: $H(\omega): y_n = H(\omega)s_n = H(\omega)e^{i\omega n}.$


$$y_{n} =\frac{s_{n+1} - s_{n-1}}{2} = \frac{e^{i\omega (n+1)} - e^{i\omega(n-1)}}{2}.$$

Значит, $$H(\omega)e^{i\omega n} = \frac{e^{i\omega (n+1)} - e^{i\omega(n-1)}}{2}.$$

$$H(\omega) = \frac{e^{i\omega (n+1)} - e^{i\omega(n-1)}}{2e^{i\omega n}} = \frac{e^{i\omega} - e^{-i\omega}}{2}.$$

Рассмотрим числитель отдельно.

$$e^{i\omega} - e^{-i\omega} = \cos\omega + i\sin\omega - \cos(-\omega) - i\sin(-\omega) = 2i\sin\omega.$$

$$H(\omega) = \frac{2i\sin\omega}{2} = i\sin\omega.$$

**Формула (средних) прямоугольников**: $$y_{n+1} = y_n + s_{n + \frac{1}{2}}, \quad y_0 = 0.$$

Рассмотрим последовательность $s_n = e^{i\omega n}.$

Найдем значение передаточной функции: $H(\omega): y_n = H(\omega)s_n = H(\omega)e^{i\omega n}.$


По определению передаточной функции: $$y_{n+1} = H(\omega)e^{i\omega (n+1)},$$ в то же время:

$$y_{n+1} = y_n + s_{n + \frac{1}{2}} = H(\omega)e^{i\omega n} + e^{i\omega(n+\frac{1}{2})}.$$

Значит, $$H(\omega)e^{i\omega (n+1)} = H(\omega)e^{i\omega n} + e^{i\omega(n+\frac{1}{2})}.$$
$$H(\omega)(e^{i\omega (n+1)} - e^{i\omega n}) = e^{i\omega (n + \frac{1}{2})}.$$
$$H(\omega) = \frac{e^{i\omega n} \cdot e^\frac{i\omega}{2}}{e^{i\omega n}(e^{i\omega} - 1)}) = \frac{e^\frac{i\omega}{2}}{e^{i\omega} - 1}.$$

Рассмотрим знаменатель отдельно.

$$e^{i\omega}-1 = \cos\omega + i\sin\omega - \cos^2\frac{\omega}{2} - \sin^2\frac{\omega}{2} =\\
= \cos(\frac{\omega}{2}\cdot2) + i\sin(\frac{\omega}{2}\cdot2) - \cos^2\frac{\omega}{2} - \sin^2\frac{\omega}{2} =\\
= \cos^2\frac{\omega}{2} - \sin^2\frac{\omega}{2} + 2i\cos\frac{\omega}{2}\sin\frac{\omega}{2} - \cos^2\frac{\omega}{2} - \sin^2\frac{\omega}{2} =\\
= 2\sin\frac{\omega}{2}(i\cos\frac{\omega}{2} - \sin\frac{\omega}{2}) =\\
= 2\sin\frac{\omega}{2}\cdot ie^\frac{\omega}{2}.$$

Таким образом,
$$H(\omega) = \frac{e^\frac{i\omega}{2}}{2\sin\frac{\omega}{2}\cdot ie^\frac{\omega}{2}} = \frac{1}{2i\sin\frac{\omega}{2}}.$$

**Формула трапеций**: $$y_{n+1} = y_n + \frac{1}{2}(s_n + s_{n+1}), \quad y_0 = 0.$$

По аналогии:
$$y_{n+1} = H(\omega)e^{i\omega n} + \frac{e^{i\omega n}+e^{i\omega (n + 1)}}{2}.$$

Значит, $$H(\omega)e^{i\omega (n+1)} = H(\omega)e^{i\omega n} + \frac{e^{i\omega n}(e^{i\omega} + 1)}{2}.$$
$$H(\omega)(e^{i\omega (n+1)} - e^{i\omega n}) = \frac{e^{i\omega n}(e^{i\omega} + 1)}{2}.$$
$$H(\omega) = \frac{e^{i\omega n}(e^{i\omega} + 1)}{2e^{i\omega n}(e^{i\omega} - 1)} = \frac{e^{i\omega} + 1}{2(e^{i\omega} - 1)}.$$

Знаменатель был вычислен выше. Найдем значение числителя.

$$e^{i\omega}+1 = \cos\omega + i\sin\omega + \cos^2\frac{\omega}{2} + \sin^2\frac{\omega}{2} =\\
= \cos(\frac{\omega}{2}\cdot2) + i\sin(\frac{\omega}{2}\cdot2) + \cos^2\frac{\omega}{2} + \sin^2\frac{\omega}{2} =\\
= \cos^2\frac{\omega}{2} - \sin^2\frac{\omega}{2} + 2i\cos\frac{\omega}{2}\sin\frac{\omega}{2} + \cos^2\frac{\omega}{2} + \sin^2\frac{\omega}{2} =\\
= 2\cos\frac{\omega}{2}(\cos\frac{\omega}{2} + i\sin\frac{\omega}{2}) =\\
= 2\cos\frac{\omega}{2}\cdot e^\frac{\omega}{2}.$$

Таким образом,
$$H(\omega) = \frac{2\cos\frac{\omega}{2}\cdot e^\frac{\omega}{2}}{4\sin\frac{\omega}{2}\cdot ie^\frac{\omega}{2}} = \frac{\cos\frac{\omega}{2}}{2i\sin\frac{\omega}{2}}.$$

**Формула Симпсона**: $$y_{n+1} = y_{n-1} + \frac{1}{3}(s_{n-1} + 4s_n + s_{n+1}), \quad y_0 = 0.$$

По аналогии:
$$y_{n+1} = H(\omega)e^{i\omega (n-1)} + \frac{1}{3}(e^{i\omega(n-1)} + 4e^{i\omega n} + e^{i\omega(n+1)}).$$

Значит, $$H(\omega)e^{i\omega (n+1)} = H(\omega)e^{i\omega (n-1)} + \frac{1}{3}(e^{i\omega(n-1)} + 4e^{i\omega n} + e^{i\omega(n+1)}).$$
$$H(\omega)(e^{i\omega (n+1)} - e^{i\omega (n - 1)}) = \frac{1}{3}(e^{i\omega(n-1)} + 4e^{i\omega n} + e^{i\omega(n+1)}).$$
$$H(\omega) = \frac{e^{i\omega n}(e^{-i\omega} + 4 + e^{i\omega})}{3e^{i\omega n}(e^{i\omega} - e^{-i\omega})} = \frac{e^{-i\omega} + 4 + e^{i\omega}}{3(e^{i\omega} - e^{-i\omega})}.$$

Рассмотрим выражение $e^{i\omega} + e^{-i\omega}.$

$$e^{i\omega} + e^{-i\omega} = \cos\omega + i\sin\omega + \cos(-\omega) + i\sin(-\omega) = 2\cos\omega.$$

Для выражения $e^{i\omega} - e^{-i\omega}$ значение вычислено выше.

Пользуясь этими равенствами:
$$H(\omega) = \frac{4 + 2\cos\omega}{6i\sin\omega} = \frac{2 + \cos\omega}{3i\sin\omega}.$$

```python
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
from math import pi, cos, sin
from cmath import exp

sns.set_theme()
random.seed()
f = np.linspace(0, 31, 500)
```

### 1) Формирование сигнала
---
Сформировать дискретный сигнал посредством дискретизации с шагом $T = 1$ непрерывного сигнала, представляющего собой линейную комбинацию косинусоид вида $A_k \cos(\omega_k t + \varphi_k)$. Частота каждой из гармоник не должна превышать $\pi$. Всего одиннадцать гармоник с упорядоченными по возрастанию частотами от 0 до $\pi$, изменяющимися с шагом $\varDelta \omega = 0.1\pi$. Амплитуды гармоник $A_k$ представляют собой целые числа со значениями от 1 до 11, определяемые случайным образом с помощью датчика равномерно распределенных случайных чисел. При необходимости нормализовать коэффициенты линейной комбинации посредством деления их на сумму полученных случайным образом амплитуд. Начальные фазы $\varphi_k$ представляют собой случайные числа в промежутке от 0 до 0.5. Дискретная последовательность должна включать в себя 32 отсчета ($N = 31$).

```python
T = 1
N = 31
k = 11
t_d = np.arange(0, N + T, T)
step = 0.01
t_a = np.arange(0, N + step, step)
omega = np.arange(0, pi+0.1, 0.1*pi)
A = list(random.randint(1, k) for i in range(k))
# нормализация
A = A / np.sum(A)
phi = list(random.random()/2 for i in range(k))
print(omega)
print(A)
print(phi)
```

### 2) Визуализация исходного и дискритезированного сигнала
---
Визуализировать исходные аналоговый и дискретизированный сигналы.

```python
def s(t):
  return sum(A[k]*cos(omega[k]*t + phi[k]) for k in range(k))
```

```python
sig = [s(t) for t in t_a]
d_sig = [s(t) for t in t_d]
plt.plot(t_a, sig, label="исходный аналоговый")
plt.stem(t_d, d_sig, 'g', markerfmt='go', label="дискретизированный сигнал")
plt.title("Исходный аналоговый и дискретизированный сигналы")
plt.xlabel("t")
plt.ylabel("Значение сигнала")
plt.legend();
```

### 3) ДПФ
---
С помощью ДПФ найти дискретные отсчеты спектра дискретного сигнала и визуализировать их.

```python
def DFT(sig, n):
  return sum(exp(-1j*2*pi*n*i/N)*sig[i] for i in range(N + 1))

spectrum = []
for i in range(N+1):
  spectrum.append(np.abs(DFT(d_sig, i)))

plt.title('Спектр дискретного сигнала')
plt.stem(t_d, np.abs(spectrum))
plt.xlabel(r'$K$')
plt.ylabel(r"$X_K$");
```

### 4)


---

Для дискретного сигнала применить линейное сглаживание по 5-ти и 9-ти точкам, представить формулу для $H(\omega)$ – передаточной функции (частотной характеристики) фильтра.

```python
def lin_smooth(sig, n, samples_num):
  return sum(sig[n + i] if (n+i >= 0 and n+i <= N) else 0 for i in range(-(samples_num//2), samples_num//2 + 1) )/samples_num

def H_lin_smooth(w, samples_num):
  return (1 + sum(2*cos(i*w) for i in range(1, samples_num//2 + 1)))/samples_num


smoothed_sig_5 = [lin_smooth(d_sig, i, 5) for i in range(N+1)]
smoothed_sig_9 = [lin_smooth(d_sig, i, 9) for i in range(N+1)]
```

8.b.1) Дискретный фильтр, соответствующий численному дифференцированию 1-го порядка.

```python
def int_rect(sig):
  res = [0]
  for i in range(1, len(sig)):
    new_val = res[-1] + s(i + 1/2)
    res.append(new_val)
  return res

def h_rect(f):
  """Передаточная функция для формулы прямоугольников"""
  numerator = 1
  denominator = 2j * np.sin(pi*f)
  h = (numerator / denominator).imag
  return h

rect_sig = int_rect(d_sig)
```

```python
def int_simps(sig):
  res = [0]
  for i in range(1, len(sig)-1, 2):
    res.append(0)
    res.append(res[-1] + (sig[i - 1] + 4*sig[i] + sig[i + 1])/3)
  return res + [0]

def h_simps(f):
  """Передаточная функция для формулы Симпсона"""
  numerator = np.cos(2*pi*f) + 2
  denominator = 3j*np.sin(2*pi*f)
  h = (numerator / denominator).imag
  return h

simps_sig = int_simps(d_sig)
```

```python
def int_trapez(sig):
  res = [0]
  for i in range(0, len(sig)-1):
    res.append(res[-1] + (sig[i] + sig[i + 1])/2)
  return res

def h_trapez(f):
  """Передаточная функция для формулы трапеций"""
  numerator = np.cos(pi*f)
  denominator = 2j*np.sin(pi*f)
  h = (numerator / denominator).imag
  return h


trapez_sig = int_trapez(d_sig)
```

Наблюдается следующее свойство: подавление крайних гармоник спектра соответствует подавлению высоких и низких частот, а подавление средних гармоник спектра — подавлению средних частот.
