## Основные теоретические положения
$H(\omega) = H(2\pi f) = \tilde H(f).$

**Формула (средних) прямоугольников**: $$y_{n+1} = y_n + s_{n + \frac{1}{2}}, \quad y_0 = 0.$$

Рассмотрим последовательность $s_n = e^{i\omega n}.$

Найдем значение передаточной функции: $H(\omega): y_n = H(\omega)s_n = H(\omega)e^{i\omega n}.$


По определению передаточной функции: $$y_{n+1} = H(\omega)e^{i\omega (n+1)},$$ в то же время:

$$y_{n+1} = y_n + s_{n + \frac{1}{2}} = H(\omega)e^{i\omega n} + e^{i\omega(n+\frac{1}{2})}.$$

Значит, $$H(\omega)e^{i\omega (n+1)} = H(\omega)e^{i\omega n} + e^{i\omega(n+\frac{1}{2})}.$$
$$H(\omega)(e^{i\omega (n+1)} - e^{i\omega n}) = e^{i\omega (n + \frac{1}{2})}.$$
$$H(\omega) = \frac{e^{i\omega n} \cdot e^\frac{i\omega}{2}}{e^{i\omega n}(e^{i\omega} - 1)}) = \frac{e^\frac{i\omega}{2}}{e^{i\omega} - 1}.$$

Рассмотрим знаменатель отдельно.

$$e^{i\omega}-1 = \cos\omega + i\sin\omega - \cos^2\frac{\omega}{2} - \sin^2\frac{\omega}{2} =\\
= \cos(\frac{\omega}{2}\cdot2) + i\sin(\frac{\omega}{2}\cdot2) - \cos^2\frac{\omega}{2} - \sin^2\frac{\omega}{2} =\\
= \cos^2\frac{\omega}{2} - \sin^2\frac{\omega}{2} + 2i\cos\frac{\omega}{2}\sin\frac{\omega}{2} - \cos^2\frac{\omega}{2} - \sin^2\frac{\omega}{2} =\\
= 2\sin\frac{\omega}{2}(i\cos\frac{\omega}{2} - \sin\frac{\omega}{2}) =\\
= 2\sin\frac{\omega}{2}\cdot ie^\frac{\omega}{2}.$$

Таким образом,
$$H(\omega) = \frac{e^\frac{i\omega}{2}}{2\sin\frac{\omega}{2}\cdot ie^\frac{\omega}{2}} = \frac{1}{2i\sin\frac{\omega}{2}}.$$

Точное значение интеграла от $e^{i\omega t}$ равно $\frac{e^{i\omega t}}{i\omega}.$
Найдем отношение вычисленного и точного значений.

$$\gamma = \frac{H(\omega)s_t}{\frac{e^{i\omega t}}{i\omega}} = \frac{e^{i\omega t}}{2i\sin\frac{\omega}{2}} \cdot \frac{i\omega}{e^{i\omega t}} = \frac{\omega}{2\sin\frac{\omega}{2}}.$$

**Формула трапеций**: $$y_{n+1} = y_n + \frac{1}{2}(s_n + s_{n+1}), \quad y_0 = 0.$$

По аналогии:
$$y_{n+1} = H(\omega)e^{i\omega n} + \frac{e^{i\omega n}+e^{i\omega (n + 1)}}{2}.$$

Значит, $$H(\omega)e^{i\omega (n+1)} = H(\omega)e^{i\omega n} + \frac{e^{i\omega n}(e^{i\omega} + 1)}{2}.$$
$$H(\omega)(e^{i\omega (n+1)} - e^{i\omega n}) = \frac{e^{i\omega n}(e^{i\omega} + 1)}{2}.$$
$$H(\omega) = \frac{e^{i\omega n}(e^{i\omega} + 1)}{2e^{i\omega n}(e^{i\omega} - 1)} = \frac{e^{i\omega} + 1}{2(e^{i\omega} - 1)}.$$

Знаменатель был вычислен выше. Найдем значение числителя.

$$e^{i\omega}+1 = \cos\omega + i\sin\omega + \cos^2\frac{\omega}{2} + \sin^2\frac{\omega}{2} =\\
= \cos(\frac{\omega}{2}\cdot2) + i\sin(\frac{\omega}{2}\cdot2) + \cos^2\frac{\omega}{2} + \sin^2\frac{\omega}{2} =\\
= \cos^2\frac{\omega}{2} - \sin^2\frac{\omega}{2} + 2i\cos\frac{\omega}{2}\sin\frac{\omega}{2} + \cos^2\frac{\omega}{2} + \sin^2\frac{\omega}{2} =\\
= 2\cos\frac{\omega}{2}(\cos\frac{\omega}{2} + i\sin\frac{\omega}{2}) =\\
= 2\cos\frac{\omega}{2}\cdot e^\frac{\omega}{2}.$$

Таким образом,
$$H(\omega) = \frac{2\cos\frac{\omega}{2}\cdot e^\frac{\omega}{2}}{4\sin\frac{\omega}{2}\cdot ie^\frac{\omega}{2}} = \frac{\cos\frac{\omega}{2}}{2i\sin\frac{\omega}{2}}.$$

$$\gamma = \frac{H(\omega)s_t}{\frac{e^{i\omega t}}{i\omega}} = \frac{\cos\frac{\omega}{2}\cdot e^{i\omega t}}{2i\sin\frac{\omega}{2}} \cdot \frac{i\omega}{e^{i\omega t}} = \frac{\omega\cos\frac{\omega}{2}}{2\sin\frac{\omega}{2}}.$$

**Формула Симпсона**: $$y_{n+1} = y_{n-1} + \frac{1}{3}(s_{n-1} + 4s_n + s_{n+1}), \quad y_0 = 0.$$

По аналогии:
$$y_{n+1} = H(\omega)e^{i\omega (n-1)} + \frac{1}{3}(e^{i\omega(n-1)} + 4e^{i\omega n} + e^{i\omega(n+1)}).$$

Значит, $$H(\omega)e^{i\omega (n+1)} = H(\omega)e^{i\omega (n-1)} + \frac{1}{3}(e^{i\omega(n-1)} + 4e^{i\omega n} + e^{i\omega(n+1)}).$$
$$H(\omega)(e^{i\omega (n+1)} - e^{i\omega (n - 1)}) = \frac{1}{3}(e^{i\omega(n-1)} + 4e^{i\omega n} + e^{i\omega(n+1)}).$$
$$H(\omega) = \frac{e^{i\omega n}(e^{-i\omega} + 4 + e^{i\omega})}{3e^{i\omega n}(e^{i\omega} - e^{-i\omega})} = \frac{e^{-i\omega} + 4 + e^{i\omega}}{3(e^{i\omega} - e^{-i\omega})}.$$

Рассмотрим выражение $e^{i\omega} + e^{-i\omega}.$

$$e^{i\omega} + e^{-i\omega} = \cos\omega + i\sin\omega + \cos(-\omega) + i\sin(-\omega) = 2\cos\omega.$$

Для выражения $e^{i\omega} - e^{-i\omega}:$

$$e^{i\omega} - e^{-i\omega} = \cos\omega + i\sin\omega - \cos(-\omega) - i\sin(-\omega) = 2i\sin\omega.$$

Пользуясь этими равенствами:
$$H(\omega) = \frac{4 + 2\cos\omega}{6i\sin\omega} = \frac{2 + \cos\omega}{3i\sin\omega}.$$

$$\gamma = \frac{H(\omega)s_t}{\frac{e^{i\omega t}}{i\omega}} = \frac{(2 + \cos\omega)\cdot e^{i\omega t}}{3i\sin\omega} \cdot \frac{i\omega}{e^{i\omega t}} = \frac{\omega (2 + \cos\omega)}{3\sin\omega}.$$

**Формула три восьмых**: $$y_{n+2} = y_{n-1} + \frac{3}{8}(s_{n-1} + 3s_n + 3s_{n+1} + s_{n+2}), \quad y_0 = 0.$$

$$y_{n+2} = H(\omega)e^{i\omega (n-1)} + \frac{3}{8}(e^{i\omega (n-1)} + 3e^{i\omega n} + 3e^{i\omega (n+1)} + e^{i\omega (n+2)}).$$

Значит, $$H(\omega)e^{i\omega (n+2)} = H(\omega)e^{i\omega (n-1)} + \frac{3}{8}(e^{i\omega (n-1)} + 3e^{i\omega n} + 3e^{i\omega (n+1)} + e^{i\omega (n+2)}).$$
$$H(\omega)(e^{i\omega (n+2)} - e^{i\omega (n - 1)}) = \frac{3}{8}(e^{i\omega (n-1)} + 3e^{i\omega n} + 3e^{i\omega (n+1)} + e^{i\omega (n+2)}).$$
$$H(\omega) = \frac{3(e^{i\omega (n-1)} + 3e^{i\omega n} + 3e^{i\omega (n+1)} + e^{i\omega (n+2)})}{8(e^{i\omega (n+2)} - e^{i\omega (n - 1)})}.$$

Разделим числитель и знаменатель дроби на $e^{i\omega (n-1)}$.

$$H(\omega) = \frac{3(1 + 3e^{i\omega} + 3e^{2i\omega} + e^{3i\omega})}{8(e^{3i\omega} - 1)}= \frac{3(1+e^{i\omega})^3}{8(e^{3i\omega} - 1)}.$$

Значение $1+e^{i\omega}$ известно. Для выражения $e^{3i\omega} - 1$ значение можно получить, подставив $\omega := 3\omega$.


Пользуясь этими равенствами:
$$H(\omega) = \frac{3\cdot(2\cos\frac{\omega}{2}\cdot e^\frac{\omega}{2})^3}{8\cdot(2\sin\frac{3\omega}{2}\cdot ie^\frac{3\omega}{2})} = \frac{3\cos^3\frac{\omega}{2}}{2i\sin\frac{3\omega}{2}}.$$


$$\gamma = \frac{H(\omega)s_t}{\frac{e^{i\omega t}}{i\omega}} = \frac{3\cos^3\frac{\omega}{2}\cdot e^{i\omega t}}{2i\sin\frac{3\omega}{2}} \cdot \frac{i\omega}{e^{i\omega t}} = \frac{3\omega \cos^3\frac{\omega}{2}}{2\sin\frac{3\omega}{2}}.$$

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from math import pi

sns.set_theme()
f = np.linspace(0, 0.5, 500)[1:]
```

### 1) Формулы прямоугольников, трапеций, Симпсона

---

Вывести формулы передаточных функций рекурсивных фильтров, соответствующих квадратурным формулам прямоугольников, трапеций и Симпсона. Построить графики передаточных функций и графики отношения вычисляемого в результате фильтрации значения к истинному. Проинтерпретировать частотные свойства полученных передаточных функций.

Формулы и их вывод приведение в разделе "Основные теоретические положения".

```python
def h_1(f):
  """Передаточная функция для формулы прямоугольников"""
  numerator = 1
  denominator = 2j * np.sin(pi*f)
  h = (numerator / denominator).imag
  return h

def h_2(f):
  """Передаточная функция для формулы трапеций"""
  numerator = np.cos(pi*f)
  denominator = 2j*np.sin(pi*f)
  h = (numerator / denominator).imag
  return h

def h_3(f):
  """Передаточная функция для формулы Симпсона"""
  numerator = np.cos(2*pi*f) + 2
  denominator = 3j*np.sin(2*pi*f)
  h = (numerator / denominator).imag
  return h

def gamma_1(f):
  """Функция отношения вычисленного значения к истонному для формулы прямоугольников"""
  numerator = pi*f
  denominator = np.sin(pi*f)
  return numerator / denominator

def gamma_2(f):
  """Функция отношения вычисленного значения к истонному для формулы трапеций"""
  numerator = np.cos(pi*f)*pi*f
  denominator = np.sin(pi*f)
  return numerator / denominator

def gamma_3(f):
  """Функция отношения вычисленного значения к истонному для формулы Симпсона"""
  numerator = (2 + np.cos(2*pi*f)) * (2*pi*f)
  denominator = 3 * np.sin(2*pi*f)
  return numerator / denominator
```

Для нахождения вычисленного значения, нужно подставить $s_n=e^{i\omega n}$ и $y_n=H(\omega )e^{iω n}$ в формулу соответствующего метода.

Формула прямоугольников.

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 7))

fig.suptitle("График передаточной функции для формулы прямоугольников")
axs[0].set_title("Полный график")
axs[0].plot(f, h_1(f))
axs[0].set_xlabel('f')
axs[0].set_ylabel(r'$\tilde H(f)$')

axs[1].set_title("Область графика (срез по оси Y)")
axs[1].plot(f, h_1(f))
axs[1].set_xlabel('f')
axs[1].set_ylabel(r'$\tilde H(f)$')
axs[1].set_ylim([-10, 0]);
```

```python
fig, axs = plt.subplots(1, 1, figsize=(12, 7))

fig.suptitle("График отношения вычисленного значения к истонному для формулы прямоугольников")
axs.plot(f, gamma_1(f))
axs.set_xlabel('f')
axs.set_ylabel(r'$\gamma(f)$');
```

При использовании формулы прямоугольников вычисленное значение при увеличении f отклоняется от истинного в большую сторону.

Формула трапеций.

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 7))

fig.suptitle("График передаточной функции для формулы трапеций")
axs[0].set_title("Полный график")
axs[0].plot(f, h_2(f))
axs[0].set_xlabel('f')
axs[0].set_ylabel(r'$\tilde H(f)$')

axs[1].set_title("Область графика (срез по оси Y)")
axs[1].plot(f, h_2(f))
axs[1].set_xlabel('f')
axs[1].set_ylabel(r'$\tilde H(f)$')
axs[1].set_ylim([-10, 1]);
```

```python
fig, axs = plt.subplots(1, 1, figsize=(12, 7))

fig.suptitle("График отношения вычисленного значения к истонному для формулы трапеций")
axs.plot(f, gamma_2(f))
axs.set_xlabel('f')
axs.set_ylabel(r'$\gamma(f)$');
```

При использовании формулы трапеций вычисленное значение с увеличением f отклоняется от истинного значения в меньшую сторону.

Формула Симпсона.

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 7))

fig.suptitle("График передаточной функции для формулы Симпсона")
axs[0].set_title("Полный график (логарифмический)")
data = [(-1 if x < 0 else 1) * np.log10(np.abs(x)) for x in h_3(f)]
axs[0].plot(f, data)
axs[0].set_xlabel('f')
axs[0].set_ylabel(r'$\tilde H(f)$')
y_ticks = list(range(-16, 2, 2))[::-1]
axs[0].set_yticks(y_ticks, [r'$-10^{' + str(abs(x)) + r'}$' for x in y_ticks])

axs[1].set_title("Область графика (срез по оси Y)")
axs[1].plot(f, h_3(f))
axs[1].set_xlabel('f')
axs[1].set_ylabel(r'$\tilde H(f)$')
axs[1].set_ylim([-10, 0]);
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 7))

fig.suptitle("График отношения вычисленного значения к истонному для формулы Симпсона")

axs[0].set_title("Полный график (логарифмический)")
axs[0].plot(f, gamma_3(f))
axs[0].set_xlabel('f')
axs[0].set_ylabel(r'$\gamma(f)$')
axs[0].set_yscale("log");

axs[1].set_title("Область графика (срез по оси Y)")
axs[1].plot(f, gamma_3(f))
axs[1].set_xlabel('f')
axs[1].set_ylabel(r'$\gamma(f)$');
axs[1].set_ylim([0, 8]);
```

### 2) Интегрирование по правилу 3/8

---


Вывести формулу передаточной функции рекурсивного фильтра для интегрирования «по правилу 3/8»:
$$ y_{n+2} = y_{n-1} + \frac{3}{8}(x_{n+2}+3x_{n+1}+3x_n+x_{n-1}) $$
Построить график передаточной функции и график отношения вычисляемого в результате фильтрации значения к истинному. Проинтерпретировать частотные свойства передаточной функции.

```python
def h_4(f):
  """Передаточная функция для формулы по правилу 3/8"""
  numerator = 3 * pow(np.cos(pi*f), 3)
  denominator = 2j * np.sin(3*pi*f)
  h = (numerator / denominator).imag
  return h

def gamma_4(f):
  """Функция отношения вычисленного значения к истонному для формулы по правилу 3/8"""
  numerator = 3*2*pi*f * pow(np.cos(pi*f), 3)
  denominator = 2*np.sin(3*pi*f)
  return numerator / denominator
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 7))

fig.suptitle("График передаточной функции для формулы по правилу 3/8")
axs[0].set_title("Полный график")
axs[0].plot(f, h_4(f))
axs[0].set_xlabel('f')
axs[0].set_ylabel(r'$\tilde H(f)$')

axs[1].set_title("Область графика (срез по оси Y)")
axs[1].plot(f, h_4(f))
axs[1].set_xlabel('f')
axs[1].set_ylabel(r'$\tilde H(f)$')
axs[1].set_ylim([-10, 10]);
```

Для частот $f$ из диапазона примерно $[0.05, 0.33]$ заметно некоторое плато, а после значения $f\approx0.33$ наблюдается заметное уменьшение передаточной функции и резкий скачок из отрицательной оси в положительную. После этого скачка высокие частоты подавляются.

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 7))

fig.suptitle("График отношения вычисленного значения к истонному для формулы Симпсона")

axs[0].set_title("Полный график")
axs[0].plot(f, gamma_4(f))
axs[0].set_xlabel('f')
axs[0].set_ylabel(r'$\gamma(f)$')

axs[1].set_title("Область графика (срез по оси Y)")
axs[1].plot(f, gamma_4(f))
axs[1].set_xlabel('f')
axs[1].set_ylabel(r'$\gamma(f)$');
axs[1].set_ylim([-5, 5]);
```

При низких частотах отклонение достаточно малое, но с ростом частоты отклонение увеличивается. После скачка вниз в значении $f\approx0.33$ вычисленное значение приближается к 0.

Передаточная функция формулы Симпсона заметно усиливает влияние высоких частот на результат. Также, функция по правилу 3/8 лучше других подавляет высокие частоты, но имеет скачок в значении ~0.33.

С увеличением частоты скорость и направление отклонения вычисленного значения от истинного меняются по-разному. Почти у всех методов отклонение идет в большую сторону. Но у формулы трапеций отклонение идёт в меньшую.


Таким образом,
1. При использовании формул трапеций и по правилу 3/8 влияние высоких частот заметно снижается.

2. При использовании формул Симпсона и прямоугольников высокие частоты имеют большее влияние на результат.

3. До точки $f≈0.2$ функции ведут себя одинаково, но после функции по формулам Сипсона и правила 3/8 отклоняются от плавного приближения к 0.
