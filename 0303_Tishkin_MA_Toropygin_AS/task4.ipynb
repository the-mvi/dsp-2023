Формула прямоугольников:

$$ y_{n+1} = y_n + s_{n+\frac{1}{2}}, y_0 = 0 $$

Пусть $s_n=e^{i \omega n}$ и $y_n=H(\omega) e^{i \omega n}$, тогда:
$$
\begin{gathered}
\left\{\begin{array}{c}
y_{n+1}=H(\omega) e^{i \omega n}+e^{i \omega\left(n+\frac{1}{2}\right)} \\
y_{n+1}=H(\omega) e^{i \omega(n+1)}
\end{array}\right. \\
H(\omega)\left(e^{i \omega n} e^{i \omega}\right)=H(\omega) e^{i \omega n}+e^{i \omega n} e^{\frac{1}{2} i \omega} \\
H(\omega)\left(e^{i \omega n} e^{i \omega}-e^{i \omega n}\right)=e^{i \omega n} e^{\frac{1}{2} i \omega} \\
H(\omega)\left(e^{i \omega}-1\right)=e^{\frac{1}{2} i \omega} \\
H(\omega)=\frac{1}{e^{\frac{i \omega}{2}}-e^{-\frac{i \omega}{2}}}=\frac{1}{2 i \sin \frac{\omega}{2}} \\
\widetilde{H}(f)=\frac{1}{2 i \sin (\pi f)}
\end{gathered}
$$

Точное значение интеграла $e^{i \omega t}$ равно $\frac{e^{i \omega t}}{i \omega}$, тогда отношение значений:
$$
\begin{gathered}
\gamma=\frac{\text { Вычисленное }}{\text { Точное }}=\frac{i \omega}{2 i \sin \frac{\omega}{2}}=\frac{\frac{\omega}{2}}{\sin \frac{\omega}{2}}=1+\frac{x^2}{24}+\frac{7 x^4}{5760}+\cdots \\
\gamma=\frac{\pi f}{\sin (\pi f)}=1+\frac{\pi^2 f^2}{6}+\frac{7 \pi^4 f^4}{360}+\cdots
\end{gathered}
$$

Формула трапеций:
$$
y_{n+1}=y_n+\frac{1}{2}\left(s_n+s_{n+1}\right), y_0=0
$$

Пусть $s_n=e^{i \omega n}$ и $y_n=H(\omega) e^{i \omega n}$, тогда:
$$
\begin{aligned}
& \left\{\begin{array}{c}
y_{n+1}=H(\omega) e^{i \omega n}+\frac{e^{i \omega n}+e^{i \omega(n+1)}}{2} \\
y_{n+1}=H(\omega) e^{i \omega(n+1)}
\end{array}\right. \\
& H(\omega)\left(e^{i \omega n} e^{i \omega}\right)=H(\omega) e^{i \omega n}+e^{i \omega n} \frac{1+e^{i \omega}}{2} \\
& H(\omega)\left(e^{i \omega}-1\right)=\frac{1+e^{i \omega}}{2} \\
& H(\omega)=\frac{1+e^{i \omega}}{2\left(e^{i \omega}-1\right)}=\frac{\cos \frac{\omega}{2}}{2 i \sin \frac{\omega}{2}} \\
&
\end{aligned}
$$

$$
\tilde{H}(f)=\frac{\cos (\pi f)}{2 i \sin (\pi f)}
$$

Точное значение интеграла $e^{i \omega t}$ равно $\frac{e^{i \omega t}}{i \omega}$, тогда отношение значений:
$$
\begin{gathered}
\gamma=\frac{\text { Вычисленное }}{\text { Точное }}=\cos \frac{\omega}{2} \frac{\frac{\omega}{2}}{\sin \frac{\omega}{2}}=1-\frac{\omega^2}{12}+\frac{\omega^4}{720}+\cdots \\
\gamma=\cos (\pi f) \frac{\pi f}{\sin (\pi f)}=1-\frac{\pi^2 \omega^2}{3}+\frac{\pi^4 \omega^4}{45}+\cdots
\end{gathered}
$$

Формула Симпсона:
$$
y_{n+1}=y_{n-1}+\frac{1}{3}\left(s_{n-1}+4 s_n+s_{n+1}\right), y_0=0
$$

Пусть $s_n=e^{i \omega n}$ и $y_n=H(\omega) e^{i \omega n}$, тогда:
$$
\begin{aligned}
& \left\{\begin{array}{c}
y_{n+1}=H(\omega) e^{i \omega(n-1)}+\frac{e^{i \omega(n-1)}+4 e^{i \omega n}+e^{i \omega(n+1)}}{3} \\
y_{n+1}=H(\omega) e^{i \omega(n+1)}
\end{array}\right. \\
& H(\omega)\left(e^{i \omega n} e^{i \omega}\right)=H(\omega) e^{i \omega n} e^{-i \omega}+e^{i \omega n} \frac{e^{-i \omega}+4+e^{i \omega}}{3} \\
& H(\omega)\left(e^{i \omega}-e^{-i \omega}\right)=\frac{e^{-i \omega}+4+e^{i \omega}}{3} \\
& H(\omega)=\frac{e^{-i \omega}+4+e^{i \omega}}{3\left(e^{i \omega}-e^{-i \omega}\right)}=\frac{\cos \omega+2}{3 i \sin \omega} \\
& \widetilde{H}(f)=\frac{\cos (2 \pi f)+2}{3 i \sin (2 \pi f)} \\
&
\end{aligned}
$$

Точное значение интеграла $e^{i \omega t}$ равно $\frac{e^{i \omega t}}{i \omega}$, тогда отношение значений:
$$
\begin{gathered}
\gamma=\frac{\text { Вычисленное }}{\text { Точное }}=\frac{(\cos \omega+2) i \omega}{3 i \sin \omega}=\frac{\cos \omega+2}{3} \cdot \frac{\omega}{\sin \omega}=1+\frac{\omega^4}{180}+\cdots \\
\gamma=\frac{\cos (2 \pi f)+2}{3} \cdot \frac{2 \pi f}{\sin (2 \pi f)}=1+\frac{4 \pi^4 f^4}{45}+\cdots
\end{gathered}
$$
