**Задача численного интегрирования** - состоит в замене исходной подинтегральной функции некой аппроксимирующей функцией. 

**Дискретный фильтр** - выполняет преобразование спектра сигнала за счет различного коэффициента передачи для разных гармонических составляющих.

**Формула прямоугольников**:
$$y_n=H(\omega)e^{i\omega n},\mspace{18mu}y_0 = 0$$

**Формула трапеций**:
$$y_{n+1} = y_n + \frac 1 2(s_n + s_{n+1}),\mspace{3mu} y_0 = 0$$

**Формула Симпсона**:
$$y_{n+1} = y_{n-1} + \frac{1}{3}(s_{n-1} + 4s_n + s_{n+1}),\mspace{3mu} y_0=0$$

**Интегрирование "по правилу 3/8"**:
$$y_{n+2} = y_{n-1} + \frac 1 8 (s_{n+2} + 3s_{n+1} + 3s_n + s_{n-1})$$

**Передаточная функция** $H(\omega)$ - периодическая функция с периодом $2\pi$; частоты дискретного сигнала не должна превышать $\pi$, из-за чего обычно рассматривается интервал $(-\pi; \pi)$ для круговой частоты $\omega$ или, в случае циклической частоты f - $(-0,5; 0,5)$:
$$H(\omega) = H(2\pi f) = \tilde{H}(f)$$

**1. Вывести формулы передаточных функций рекурсивных фильтров, соответствующих квадратурным формулам прямоугольников, трапеций и Симпсона:**

Пусть $s_n=e^{i\omega n}, \mspace{3mu} y_n=H(\omega)e^{i\omega n}\mspace{18mu}(1)$

**Формула прямоугольников имеет следующий вид:**
$$y_{n+1} = y_n + s_{n + \frac{1}{2}},\mspace{3mu}y_0 = 0$$
Тогда, используя (1) получим:

$$\left\{\begin{matrix}
y_{n+1} = H(\omega)e^{i\omega (n + 1)}\mspace{70mu} (2.1)
\\y_{n+1} = H(\omega)e^{i\omega n} + e^{i\omega (n + 0.5)}\mspace{18mu}(2.2)
\end{matrix}\right.$$

где (2.1) - точное значение, (2.2) - вычисленное.

Приравняем правые части выражений (2.1) и (2.2):
$$H(\omega)e^{i\omega (n + 1)} = H(\omega)e^{i\omega n} + e^{i\omega (n + 0.5)}$$
Разложим как:
$$H(\omega)e^{i\omega n} * e^{i\omega } = H(\omega)e^{i\omega n} + e^{i\omega n} * e^{0.5 i\omega}$$
И поделим на общий множитель $e^{i\omega n}$ правую и левую часть:
$$H(\omega)e^{i\omega} = H(\omega) + e^{0.5 i\omega}$$
$$H(\omega)(e^{i\omega} - 1) =  e^{0.5 i\omega}$$
$$H(\omega) = \frac{e^{0.5 i\omega}}{(e^{i\omega} - 1)} \mspace{18mu} (2.3)$$

По формулам Эйлера: $sin(x) = \frac{e^{ix}-e^{-ix}}{2i}\Rightarrow$
(2.3) $= H(\omega)=\frac{1}{e^{0.5i\omega} - e^{-0.5i\omega}} \Rightarrow H(\omega)=\frac{1}{2i sin(\frac{\omega}{2})}$

$\omega = 2\pi f \Rightarrow \tilde{H}(f) = \frac{1}{2i sin(\pi f}$
Отметим, что точное значение интеграла от $e^{i\omega t} = \frac{i\omega t}{i \omega}$

Отношение вычисленного и точного значений:
$$\gamma = \frac{вычисленное}{точное} = \frac{i\omega}{2isin(\frac{\omega}{2})} = \frac{\frac{\omega}{2}}{sin(\frac{\omega}{2})}=1 + \frac{\omega^2}{24}+\frac{7\omega^4}{5760}+...$$
$$\tilde \gamma = \frac {\text {Вычисленное}}{\text {Точное}} = \frac {\pi f}{\sin {\pi f}} = 1 + \frac {\pi^2 f^2}{6} + \frac {7 \pi^4 f^4}{360} + ...$$

**Формула трапеций имеет следующий вид:**
$$y_{n+1} = y_n + \frac 1 2(s_n + s_{n+1}),\mspace{3mu} y_0 = 0$$
Воспользуемся (1):
$$\left\{\begin{matrix}
y_{n+1} = H(\omega)e^{i\omega (n + 1)}\mspace{128mu} (3.1)
\\y_{n+1} = H(\omega)e^{i\omega n} + \frac{1}{2}(e^{i\omega n} + e^{i\omega (n+1)})\mspace{18mu}(3.2)
\end{matrix}\right.$$
где (3.1) - точное значение, (3.2) - вычисленное.

Приравняем правые части выражений (3.1) и (3.2):
$$H(\omega)e^{i \omega (n + 1)} = H(\omega)e^{i \omega n} + \frac{1}{2}(e^{i \omega n} + e^{i \omega (n + 1)})$$
Разложим как:
$$H(\omega)e^{i\omega n} * e^{i\omega } = H(\omega)e^{i\omega n} + \frac{1}{2}(e^{i\omega n} + e^{i\omega n} * e^{i\omega})$$
И поделим на общий множитель $e^{i\omega n}$ правую и левую часть:
$$H(\omega)e^{i\omega} = H(\omega) + \frac{1}{2}(1 + e^{i\omega})$$
$$H(\omega)(e^{i\omega} - 1) =  \frac{1}{2}(1 + e^{i\omega})$$
$$H(\omega) = \frac{\frac{1}{2}(1 + e^{i\omega})}{(e^{i\omega} - 1)} \mspace{18mu} = \frac{cos(\frac{\omega}{2})}{2 i sin(\frac{\omega}{2})}$$

$\omega = 2\pi f \Rightarrow \tilde{H}(f) = \frac{cos(\pi f)}{2i sin(\pi f}$
Отметим, что точное значение интеграла от $e^{i\omega t} = \frac{i\omega t}{i \omega}$

Отношение вычисленного и точного значений:
$$\gamma = \frac{вычисленное}{точное} = cos(\frac{\omega}{2})* \frac{i\omega}{2i sin(\frac{\omega}{2})}=1 - \frac{\omega^2}{12} + \frac{\omega^4}{720}+...$$
$$\tilde \gamma = \frac {\text {Вычисленное}}{\text {Точное}} =  \cos (\pi f) \frac {\pi f}{\sin (\pi f)} = 1 - \frac {\pi^2 f^2}{3} + \frac {\pi^4 f^4}{45} + ...$$

**Формула Симпсона имеет вид:**
$$y_{n+1} = y_{n-1} + \frac{1}{3}(s_{n-1} + 4s_n + s_{n+1}),\mspace{3mu} y_0=0$$

Воспользуемся (1):
$$\left\{\begin{matrix}
y_{n+1} = H(\omega)e^{i\omega (n + 1)}\mspace{220mu} (4.1)
\\y_{n+1} = H(\omega)e^{i\omega (n-1)} + \frac{1}{3}(e^{i\omega (n -1)}+4e^{i\omega n} + e^{i\omega (n+1)})\mspace{18mu}(4.2)
\end{matrix}\right.$$

где (4.1) - точное значение, (4.2) - вычисленное.

Приравняем правые части выражений (4.1) и (4.2):
$$H(\omega)e^{i \omega (n + 1)} = H(\omega)e^{i \omega (n-1)} + \frac{1}{3}(e^{i \omega (n-1)} + 4e^{i \omega n}+e^{i \omega (n+1)})$$
Разложим как:
$$H(\omega)e^{i\omega n} * e^{i\omega } = H(\omega)e^{i\omega n} * e^{-i\omega} + \frac{1}{3}(e^{i \omega n}*e^{-i \omega} + 4e^{i \omega n}+e^{i \omega n}*e^{i \omega})$$
И поделим на общий множитель $e^{i\omega n}$ правую и левую часть:
$$H(\omega)e^{i\omega} = H(\omega)e^{-i\omega} + \frac{1}{3}(e^{-i \omega} + 4+e^{i \omega})$$
$$H(\omega)(e^{i\omega} - e^{-i\omega}) =  \frac{1}{3}(e^{-i \omega} + 4+e^{i \omega})$$

$$H(\omega) = \frac{\frac{1}{3}(e^{-i\omega}+4+ e^{i\omega})}{(e^{i\omega} - e^{-i\omega})} \mspace{18mu} = \frac{cos\omega+2}{3 i sin \omega}$$

$\omega = 2\pi f \Rightarrow \tilde{H}(f) = \frac{cos(2\pi f) +2}{3i sin(2\pi f)}$
Отметим, что точное значение интеграла от $e^{i\omega t} = \frac{i\omega t}{i \omega}$

Отношение вычисленного и точного значений:
$$\gamma = \frac{вычисленное}{точное} = \frac{(cos\omega + 2)i\omega}{3i sin(\omega)}=\frac{cos(\omega)+2}{3}*\frac{\omega}{sin(\omega)}=1 + \frac{\omega^4}{180} + ...$$
$$\tilde \gamma = \frac {\text {Вычисленное}}{\text {Точное}} = \frac {\cos (2 \pi f) + 2} {3} * \frac {2 \pi f}{\sin (2 \pi f)} = 1 + \frac {4 \pi^4 f^4}{45} + ...$$


Пропишем выведенные формулы:

```python
import numpy as np
import matplotlib.pyplot as plt
import math

#transf funcs:
def transfer_H_rect(f):
    return (1 / 3j * np.sin(math.pi * f))

def transfer_H_trap(f):
    return (np.cos(math.pi * f) / (2j * np.sin(math.pi * f)))

def transfer_H_simpson(f):
    return ((np.cos(2 * math.pi * f) + 2) / (3j * np.sin(2 * math.pi * f)))

#gamma - difference:
def gamma_rect(f):
    return (math.pi*f) / (np.sin(math.pi*f))

def gamma_trap(f):
    return (np.cos(math.pi*f))*((math.pi*f) / (np.sin(math.pi*f)))

def gamma_simpson(f):
    return ((np.cos(2*math.pi*f)+2)/3) * ((2*math.pi*f)/(np.sin(2*math.pi*f)))
```

Построим графики передаточных функций:

```python
fr = np.arange(0.001, 0.5, 0.001)

plt.figure(figsize=(10, 5))
plt.plot(fr, transfer_H_rect(fr).imag, 'r-', label='Rectangular eq',)
plt.plot(fr, transfer_H_trap(fr).imag, 'b-', label='Trapeze eq')
plt.plot(fr, transfer_H_simpson(fr).imag, 'y-', label='Simpson eq')
plt.xlabel('f')
plt.ylabel('tildaH')
plt.axis((0, 0.5, -10, 3))
plt.grid()


plt.legend()
plt.show()
```

Графики отношений вычисляемого значения в результате фильтрации к истинному значению:

```python
plt.figure(figsize=(10, 5))
plt.plot(fr, gamma_rect(fr), 'r-', label='Rectangular eq',)
plt.plot(fr, gamma_trap(fr), 'b-', label='Trapeze eq')
plt.plot(fr, gamma_simpson(fr), 'y-', label='Simpson eq')
plt.xlabel('f')
plt.ylabel('tilda Gamma')
plt.axis((0, 0.5, -1, 3))
plt.grid()


plt.legend()
plt.show()
```

Из получившихся графиков можно сделать вывод, что если подинтегральная функция изменчива (множетсво высокочастотных компонентов), то для формулы прямоугольников и Симпсона в выходном значении роль таких компонентов будет усиливаться, а при использовании формулы трапеций высокочастотные компоненты подавляются на выходе.  

**2.Вывести формулу передаточной функции рекурсивного фильтра для интегрирования «по правилу 3/8»**
$$y_{n+2} = y_{n-1} + \frac 1 8 (s_{n+2} + 3s_{n+1} + 3s_n + s_{n-1})$$

Пусть используем (1):

$$\left\{\begin{matrix}
y_{n+1} = H(\omega)e^{i\omega (n + 2)}\mspace{290mu} (5.1)
\\y_{n+1} = H(\omega)e^{i\omega (n-1)} + \frac{1}{8}(e^{i\omega (n +2)}+3e^{i\omega (n+1)} + 3e^{i\omega n}+e^{i\omega (n-1)})\mspace{18mu}(5.2)
\end{matrix}\right.$$

где (5.1) - точное значение, (5.2) - вычисленное.



Приравняем правые части выражений (5.1) и (5.2):
$$H(\omega)e^{i \omega (n + 2)} = H(\omega)e^{i \omega (n-1)} + \frac{1}{8}(e^{i \omega (n+2)} + 3e^{i \omega (n+1)}+3e^{i \omega n}+e^{i \omega (n-1)})$$
Разложим как:
$$H(\omega)(e^{i \omega n}*e^{2 i \omega}) = H(\omega)(e^{i \omega n}*e^{-i \omega})  + \frac 1 8 (e^{i \omega n}*e^{2 i \omega } + 3e^{i \omega n}* e^{ i \omega} + 3e^{i \omega n} + e^{i \omega n} * e^{-i \omega})$$
И поделим на общий множитель $e^{i\omega n}$ правую и левую часть:
$$H(\omega)(e^{2 i \omega} - e^{-i \omega}) = \frac 1 8 (e^{2 i \omega } + 3e^{ i \omega} + 3 + e^{-i \omega})$$

$$H(\omega) = \frac {e^{2i \omega} + 3e^{i \omega} + 3 + e^{- i \omega}}{8 (e^{2 i \omega} - e^{- i \omega})}$$

$$H(\omega) = \frac {e^{2i \omega} + 3e^{i \omega} + 3 + e^{- i \omega}}{8 (e^{2 i \omega} - e^{- i \omega})} * \frac {e^{- \frac {i\omega}{2}}}{e^{- \frac {i\omega}{2}}} = \frac {e^{\frac {3i\omega}{2}} + 3e^{\frac {i\omega}{2}} + 3e^{- \frac {i\omega}{2}} + e^{- \frac {3i\omega}{2}}}{8 (e^{\frac {3i\omega}{2}} -\ e^{- \frac {3i\omega}{2}})} = \frac {2 \cos {(\frac {3\omega} 2)} + 6 \cos{(\frac \omega 2)}}{16 i \sin (\frac {3\omega} 2)}$$

$$\omega = 2πf \Rightarrow  \tilde H(f) = \frac {\cos {(3 \pi f)} + 3 \cos{(\pi f)}}{8 i \sin (3 \pi f)}$$

Точное значение интеграла от $e^{i \omega t} = \frac {e^{i \omega t}} {i \omega}$:

$$\gamma = \frac {\text {Вычисленное}}{\text {Точное}} = \frac {(\cos {(\frac {3\omega} 2)} + 3 \cos{(\frac \omega 2)})} {12}  * \frac {\frac {3\omega}{2}}{\sin (\frac {3\omega}{2})} = \frac 1 3 + \frac {\omega^4}{240} + ...$$

$$\tilde \gamma = \frac {\text {Вычисленное}}{\text {Точное}} = \frac {(\cos {(3 \pi f)} + 3 \cos{(\pi f)})} {12}  * \frac {3 \pi f}{\sin (3 \pi f)} = \frac 1 3 + \frac {\pi^4 f^4}{15} + ...$$

Пропишем выведенные формулы:

```python
#Передаточная функция, соответствующая формуле 3/8
def transfer_H_38(f):
    return ((np.cos(3 * math.pi * f) + 3 * np.cos(math.pi * f)) / (8j * np.sin(3 * math.pi * f)))

#отношение:
def gamma_38(f):
    return 1/12 * (np.cos(3 * math.pi * f) + 3 * np.cos(math.pi * f)) * (3 * math.pi * f) / (np.sin(3 * math.pi * f))
```

Выведем график передаточной функции:

```python
plt.figure(figsize=(10, 5))
plt.plot(fr, transfer_H_38(fr).imag, 'g-', label='3/8 eq',)

plt.xlabel('f')
plt.ylabel('tilda H')
plt.axis((0, 0.5, -30, 15))
plt.grid()

plt.legend()
plt.show()
```

Выведем график отношения вычисленного после фильтрации значения к истинному:

```python
plt.figure(figsize=(10, 5))
plt.plot(fr, gamma_38(fr), 'g-', label='3/8 eq',)

plt.xlabel('f')
plt.ylabel('tilda H')
plt.axis((0, 0.5, -25, 45))
plt.grid()

plt.legend()
plt.show()
```

По данным графикам можно заметить, что для данного рекурсивного фильтра в интервале $[0; 0,2)$ происходит подавление сигнала, а в приблихительном интервале $(0,31; 0,36)$ наблюдаются скачки, которые отражают искажение сигнала относительно влияния высокочастотных компонентов. На остальных интервалах графика проходит полоса пропускания фильтра, где сигнал не искажается.
